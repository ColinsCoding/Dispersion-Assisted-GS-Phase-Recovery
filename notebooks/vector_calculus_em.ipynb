# Vector Calculus → Griffiths E&M
Arithmetic → vectors → derivatives → ∇ → Maxwell. Every step shown.

```python
from sympy import *
from sympy.vector import CoordSys3D, gradient, divergence, curl, laplacian
from IPython.display import display, Markdown

init_printing(use_latex='mathjax')
C = CoordSys3D('C')
x, y, z = symbols('x y z', real=True)
r, theta, phi = symbols('r theta phi', positive=True)

def step(label, expr=None):
    display(Markdown(f'**{label}**'))
    if expr is not None:
        display(expr)

def section(title):
    display(Markdown(f'---\n## {title}'))
```

## §1 — Arithmetic → Algebra: What a Vector Is

```python
section('Vectors from scratch')

display(Markdown(r'''
A **scalar** is one number: mass $m = 5$ kg, temperature $T = 300$ K.

A **vector** is a list of numbers with direction:
$$\mathbf{A} = A_x\hat{x} + A_y\hat{y} + A_z\hat{z} = \begin{pmatrix}A_x\\A_y\\A_z\end{pmatrix}$$
'''))

Ax, Ay, Az, Bx, By, Bz = symbols('A_x A_y A_z B_x B_y B_z', real=True)
A = Matrix([Ax, Ay, Az])
B = Matrix([Bx, By, Bz])

step('Vector A =', A)

# Magnitude
mag_A = sqrt(Ax**2 + Ay**2 + Az**2)
step('|A| = √(Ax²+Ay²+Az²) =', mag_A)

# Unit vector
step('Â = A/|A| (unit vector, magnitude=1)')

# Addition
step('A + B =', A + B)

# Dot product
dot = A.dot(B)
step('A · B = AxBx + AyBy + AzBz =', dot)
step('= |A||B|cos(θ)  →  zero if perpendicular')

# Cross product
cross = A.cross(B)
step('A × B =', cross)
step('|A×B| = |A||B|sin(θ)  →  zero if parallel')
step('Direction: right-hand rule, perpendicular to both A and B')

# BAC-CAB
step('BAC-CAB identity: A×(B×C) = B(A·C) − C(A·B)')
```

## §2 — Derivatives: Slope, Then Partial Derivatives

```python
section('From slope to partial derivatives')

display(Markdown(r'''
**Slope** (1D): $f'(x) = \lim_{h\to 0}\frac{f(x+h)-f(x)}{h}$

**Partial derivative**: hold all but one variable fixed.
$$\frac{\partial f}{\partial x} \equiv \text{slope in x-direction only}$$
'''))

f = x**2 * y + z * sin(x)
step('f(x,y,z) =', f)

dfx = diff(f, x)
dfy = diff(f, y)
dfz = diff(f, z)
step('∂f/∂x =', dfx)
step('∂f/∂y =', dfy)
step('∂f/∂z =', dfz)

# Total differential
dx_sym, dy_sym, dz_sym = symbols('dx dy dz')
df_total = dfx*dx_sym + dfy*dy_sym + dfz*dz_sym
step('Total differential df = (∂f/∂x)dx + (∂f/∂y)dy + (∂f/∂z)dz =', df_total)

# Chain rule
t_sym = symbols('t')
x_t = Function('x')(t_sym)
y_t = Function('y')(t_sym)
step('Chain rule: df/dt = (∂f/∂x)(dx/dt) + (∂f/∂y)(dy/dt) + ...')
```

## §3 — The Gradient ∇f: Steepest Ascent

```python
section('Gradient')

display(Markdown(r'''
$$\nabla f = \frac{\partial f}{\partial x}\hat{x} + \frac{\partial f}{\partial y}\hat{y} + \frac{\partial f}{\partial z}\hat{z}$$

- Points in direction of **steepest increase** of $f$
- Magnitude $|\nabla f|$ = rate of increase in that direction
- Perpendicular to level surfaces (equipotentials)
'''))

f2 = x**2 + y**2 + z**2  # r²
step('f = x²+y²+z² = r²:', f2)
grad_f2 = Matrix([diff(f2,x), diff(f2,y), diff(f2,z)])
step('∇f =', grad_f2)
step('= 2r̂  →  gradient of r² points radially outward')

# Gradient of 1/r — crucial for Griffiths
r_mag = sqrt(x**2 + y**2 + z**2)
f3 = 1/r_mag
grad_f3 = Matrix([diff(f3,x), diff(f3,y), diff(f3,z)])
grad_f3_simplified = Matrix([simplify(c) for c in grad_f3])
step('f = 1/r =', f3)
step('∇(1/r) =', grad_f3_simplified)
step('= −r̂/r²  →  points INWARD, magnitude falls as 1/r²')
step('This is exactly the electric field direction from a point charge')

# Directional derivative
n1, n2, n3 = symbols('n_1 n_2 n_3', real=True)
n_hat = Matrix([n1, n2, n3])
dir_deriv = grad_f2.dot(n_hat)
step('Directional derivative ∇f · n̂ =', dir_deriv)
step('= rate of change of f in direction n̂')
```

## §4 — Divergence ∇·F: Sources and Sinks

```python
section('Divergence')

display(Markdown(r'''
$$\nabla \cdot \mathbf{F} = \frac{\partial F_x}{\partial x} + \frac{\partial F_y}{\partial y} + \frac{\partial F_z}{\partial z}$$

- **Positive divergence**: net outflow (source — like positive charge)
- **Negative divergence**: net inflow (sink — like negative charge)
- **Zero divergence**: no sources/sinks (field lines continuous)
'''))

Fx = x**2
Fy = y**2
Fz = z**2
div_F = diff(Fx,x) + diff(Fy,y) + diff(Fz,z)
step('F = x²x̂ + y²ŷ + z²ẑ')
step('∇·F = ∂Fx/∂x + ∂Fy/∂y + ∂Fz/∂z =', div_F)

# Uniform field: zero divergence
step('Uniform field F = E₀x̂: ∇·F =', diff(symbols('E_0'), x))
step('= 0  →  no sources in empty space')

# Gauss divergence theorem
display(Markdown(r'''
**Divergence theorem** (Gauss):
$$\oint_S \mathbf{F}\cdot d\mathbf{a} = \int_V (\nabla\cdot\mathbf{F})\,d\tau$$

Flux through closed surface = integral of divergence over enclosed volume.
This is **Gauss's law** for E:
$$\oint \mathbf{E}\cdot d\mathbf{a} = \frac{Q_{enc}}{\varepsilon_0} \iff \nabla\cdot\mathbf{E} = \frac{\rho}{\varepsilon_0}$$
'''))

# Divergence of r̂/r²
display(Markdown(r'''
**Key result** (Griffiths Example 1.16):
$$\nabla\cdot\left(\frac{\hat{r}}{r^2}\right) = 4\pi\delta^3(\mathbf{r})$$

Looks zero everywhere except origin — but integral over any sphere = $4\pi$.
This is why point charges produce $\nabla\cdot\mathbf{E} = \rho/\varepsilon_0$ with $\rho = q\delta^3(\mathbf{r})$.
'''))
```

## §5 — Curl ∇×F: Rotation and Circulation

```python
section('Curl')

display(Markdown(r'''
$$\nabla\times\mathbf{F} = \begin{vmatrix}\hat{x}&\hat{y}&\hat{z}\\
\partial_x&\partial_y&\partial_z\\ F_x&F_y&F_z\end{vmatrix}
= \left(\frac{\partial F_z}{\partial y}-\frac{\partial F_y}{\partial z}\right)\hat{x}
+ \left(\frac{\partial F_x}{\partial z}-\frac{\partial F_z}{\partial x}\right)\hat{y}
+ \left(\frac{\partial F_y}{\partial x}-\frac{\partial F_x}{\partial y}\right)\hat{z}$$

- **Nonzero curl**: field circulates (like B field around a wire)
- **Zero curl**: conservative field (work path-independent, has potential)
'''))

# B field from infinite wire: B = (μ₀I/2πs) φ̂  — in cylindrical, but show curl
# In Cartesian: B = μ₀I/(2π) * (-y, x, 0)/(x²+y²)
mu0, I_wire = symbols('mu_0 I', positive=True)
s2 = x**2 + y**2
Bx_wire = -mu0*I_wire/(2*pi) * y/s2
By_wire =  mu0*I_wire/(2*pi) * x/s2
Bz_wire = S.Zero

curl_B_z = diff(By_wire,x) - diff(Bx_wire,y)
step('B from infinite wire (Cartesian):')
step('∇×B in z-component =', simplify(curl_B_z))
step('= 0 away from wire  (current only at origin → delta function there)')

# Curl of gradient = 0 always
f_pot = x**2*y + y*z
grad_f = Matrix([diff(f_pot,x), diff(f_pot,y), diff(f_pot,z)])
curl_grad = Matrix([
    diff(grad_f[2],y) - diff(grad_f[1],z),
    diff(grad_f[0],z) - diff(grad_f[2],x),
    diff(grad_f[1],x) - diff(grad_f[0],y)
])
step('∇×(∇f) = (always zero):', curl_grad)

# Stokes theorem
display(Markdown(r'''
**Stokes theorem**:
$$\oint_C \mathbf{F}\cdot d\mathbf{l} = \int_S (\nabla\times\mathbf{F})\cdot d\mathbf{a}$$

Line integral around closed loop = flux of curl through any surface bounded by that loop.
This is **Faraday's law**:
$$\oint \mathbf{E}\cdot d\mathbf{l} = -\frac{d\Phi_B}{dt} \iff \nabla\times\mathbf{E} = -\frac{\partial\mathbf{B}}{\partial t}$$
'''))
```

## §6 — Laplacian ∇²: Diffusion and Potential

```python
section('Laplacian')

display(Markdown(r'''
$$\nabla^2 f = \frac{\partial^2 f}{\partial x^2} + \frac{\partial^2 f}{\partial y^2} + \frac{\partial^2 f}{\partial z^2}$$

= divergence of gradient = "how much $f$ at a point differs from average of surroundings"
'''))

f_lap = x**2 + y**2 - 2*z**2
lap_f = diff(f_lap,x,2) + diff(f_lap,y,2) + diff(f_lap,z,2)
step('f = x²+y²−2z²:', f_lap)
step('∇²f =', lap_f)
step('= 0  →  f is harmonic (satisfies Laplace equation)')

# Laplace equation in E&M
display(Markdown(r'''
**Laplace equation**: $\nabla^2 V = 0$ in charge-free region

**Poisson equation**: $\nabla^2 V = -\rho/\varepsilon_0$ with charge

**Uniqueness theorem**: V is uniquely determined by boundary conditions alone.
'''))

# Laplacian in spherical (Griffiths Eq. 1.71)
display(Markdown(r'''
**Spherical coordinates** (essential for Griffiths Ch. 2–4):
$$\nabla^2 V = \frac{1}{r^2}\frac{\partial}{\partial r}\left(r^2\frac{\partial V}{\partial r}\right)
+ \frac{1}{r^2\sin\theta}\frac{\partial}{\partial\theta}\left(\sin\theta\frac{\partial V}{\partial\theta}\right)
+ \frac{1}{r^2\sin^2\theta}\frac{\partial^2 V}{\partial\phi^2}$$
'''))

# Verify: ∇²(1/r) = 0 for r≠0
r_sym = sqrt(x**2+y**2+z**2)
lap_1r = diff(1/r_sym,x,2)+diff(1/r_sym,y,2)+diff(1/r_sym,z,2)
step('∇²(1/r) for r≠0:', simplify(lap_1r))
step('= 0  →  Coulomb potential satisfies Laplace equation everywhere except origin')
step('At origin: ∇²(1/r) = −4πδ³(r)  →  Poisson equation with point charge')
```

## §7 — Coordinate Systems: Spherical and Cylindrical

```python
section('Curvilinear Coordinates')

display(Markdown(r'''
**Spherical** $(r,\theta,\phi)$:  
$x = r\sin\theta\cos\phi$, $y = r\sin\theta\sin\phi$, $z = r\cos\theta$

$$\nabla f = \frac{\partial f}{\partial r}\hat{r} + \frac{1}{r}\frac{\partial f}{\partial\theta}\hat{\theta} + \frac{1}{r\sin\theta}\frac{\partial f}{\partial\phi}\hat{\phi}$$

$$\nabla\cdot\mathbf{A} = \frac{1}{r^2}\frac{\partial(r^2 A_r)}{\partial r} + \frac{1}{r\sin\theta}\frac{\partial(\sin\theta\, A_\theta)}{\partial\theta} + \frac{1}{r\sin\theta}\frac{\partial A_\phi}{\partial\phi}$$

Volume element: $d\tau = r^2\sin\theta\,dr\,d\theta\,d\phi$

---

**Cylindrical** $(s,\phi,z)$:  
$x = s\cos\phi$, $y = s\sin\phi$, $z = z$

$$\nabla f = \frac{\partial f}{\partial s}\hat{s} + \frac{1}{s}\frac{\partial f}{\partial\phi}\hat{\phi} + \frac{\partial f}{\partial z}\hat{z}$$

Volume element: $d\tau = s\,ds\,d\phi\,dz$
'''))

# Solid angle
display(Markdown(r'''
**Solid angle**: $d\Omega = \sin\theta\,d\theta\,d\phi$, total sphere = $4\pi$ sr

$$\oint d\Omega = \int_0^{2\pi}d\phi\int_0^\pi\sin\theta\,d\theta = 4\pi$$
'''))
phi_sym = symbols('phi_s')
solid = integrate(integrate(sin(theta), (theta, 0, pi)), (phi_sym, 0, 2*pi))
step('∫∫ sinθ dθ dφ over full sphere =', solid)
```

## §8 — Maxwell's Equations: All Four from Vector Calculus

```python
section('Maxwell from ∇')

display(Markdown(r'''
Every Maxwell equation is one of: divergence, curl, or their integral form via Gauss/Stokes.

| Law | Differential | Integral | Vector op |
|---|---|---|---|
| Gauss E | $\nabla\cdot\mathbf{E}=\rho/\varepsilon_0$ | $\oint\mathbf{E}\cdot d\mathbf{a}=Q_{enc}/\varepsilon_0$ | divergence |
| Gauss B | $\nabla\cdot\mathbf{B}=0$ | $\oint\mathbf{B}\cdot d\mathbf{a}=0$ | divergence |
| Faraday | $\nabla\times\mathbf{E}=-\partial\mathbf{B}/\partial t$ | $\oint\mathbf{E}\cdot d\mathbf{l}=-d\Phi_B/dt$ | curl |
| Ampere-Maxwell | $\nabla\times\mathbf{B}=\mu_0\mathbf{J}+\mu_0\varepsilon_0\partial\mathbf{E}/\partial t$ | $\oint\mathbf{B}\cdot d\mathbf{l}=\mu_0 I_{enc}+\mu_0\varepsilon_0 d\Phi_E/dt$ | curl |
'''))

# Derive wave equation from Faraday + Ampere in vacuum
display(Markdown(r'''
**Wave equation from Maxwell** (vacuum, $\rho=0$, $\mathbf{J}=0$):

Take curl of Faraday:
$$\nabla\times(\nabla\times\mathbf{E}) = -\frac{\partial}{\partial t}(\nabla\times\mathbf{B}) = -\mu_0\varepsilon_0\frac{\partial^2\mathbf{E}}{\partial t^2}$$

Use vector identity $\nabla\times(\nabla\times\mathbf{E}) = \nabla(\nabla\cdot\mathbf{E}) - \nabla^2\mathbf{E}$
and Gauss ($\nabla\cdot\mathbf{E}=0$ in vacuum):

$$\boxed{\nabla^2\mathbf{E} = \mu_0\varepsilon_0\frac{\partial^2\mathbf{E}}{\partial t^2}}$$

This is the wave equation with $c = 1/\sqrt{\mu_0\varepsilon_0} = 2.998\times10^8$ m/s.
'''))

epsilon0, mu0 = symbols('epsilon_0 mu_0', positive=True)
c_light = 1/sqrt(mu0*epsilon0)
step('c = 1/√(μ₀ε₀) =', c_light)
step('Numerically:', '2.998 × 10⁸ m/s  →  light speed falls out of vector calculus')
```

## §9 — Coulomb's Law and Electric Potential

```python
section('Coulomb → Potential → Gradient')

q, Q, epsilon0_s, r_dist = symbols('q Q epsilon_0 r', positive=True)

# Coulomb
F_coulomb = q*Q / (4*pi*epsilon0_s * r_dist**2)
step('Coulomb force (magnitude): F = qQ/(4πε₀r²) =', F_coulomb)

# E field from point charge
E_point = Q / (4*pi*epsilon0_s * r_dist**2)
step('Electric field E = Q/(4πε₀r²) r̂ =', E_point, '× r̂')

# Potential
V_point = Q / (4*pi*epsilon0_s * r_dist)
step('Potential V = Q/(4πε₀r) =', V_point)
step('E = −∇V  →  verify: −dV/dr =', -diff(V_point, r_dist))
step('= Q/(4πε₀r²)  ✓ matches Coulomb')

# Superposition
display(Markdown(r'''
**Superposition**: potentials add as scalars (easier than vector addition of E fields)
$$V(\mathbf{r}) = \frac{1}{4\pi\varepsilon_0}\sum_i \frac{q_i}{|\mathbf{r}-\mathbf{r}_i|}$$
then $\mathbf{E} = -\nabla V$ — one vector operation at the end.
'''))

# Energy stored in field
display(Markdown(r'''
**Energy in electric field** (Griffiths Eq. 2.45):
$$W = \frac{\varepsilon_0}{2}\int E^2\,d\tau$$

Energy density: $u = \frac{\varepsilon_0}{2}E^2 + \frac{1}{2\mu_0}B^2$
— field carries energy even in vacuum. This is measurable (radiation pressure).
'''))
```

## §10 — Hypersonic Sensing: Plasma Phase Shift via ∇

```python
section('Plasma Sheath — Vector Calculus Connection')

display(Markdown(r'''
Reentry plasma has electron density $n_e(\mathbf{r})$ — a **scalar field**.
Plasma frequency: $\omega_p(\mathbf{r}) = \sqrt{n_e e^2/(\varepsilon_0 m_e)}$

Refractive index field:
$$n(\mathbf{r},\omega) = \sqrt{1 - \frac{\omega_p(\mathbf{r})^2}{\omega^2}}$$

**Wave equation in inhomogeneous medium**:
$$\nabla^2\mathbf{E} + \frac{\omega^2}{c^2}n^2(\mathbf{r})\mathbf{E} = 0$$

This is the **Helmholtz equation** — same PDE as quantum mechanics
($\nabla^2\psi + k^2\psi = 0$) and acoustics.

**Phase accumulated** through plasma slab:
$$\phi(\omega) = \frac{\omega}{c}\int_0^L n(z,\omega)\,dz \approx \frac{\omega L}{c} - \frac{\omega_p^2 L}{2c\omega}$$

**TDGSA sensor**: two optical wavelengths through plasma → two phase measurements
→ GS retrieval reconstructs $n_e(z)$ profile.
Sell to SpaceX: non-contact electron density tomography at Mach 10+, no RF blackout.
'''))

omega_sym, omega_p, L_plasma = symbols('omega omega_p L', positive=True)
n_plasma = sqrt(1 - omega_p**2/omega_sym**2)
step('n(ω) =', n_plasma)

phi_plasma = omega_sym/3e8 * L_plasma * n_plasma
step('Phase φ(ω) = ωL·n(ω)/c ≈', series(phi_plasma, omega_p, 0, 3))

# Group delay
dphi = diff(omega_sym * L_plasma * n_plasma / 3e8, omega_sym)
step('Group delay dφ/dω =', simplify(dphi))
step('Dispersive: different frequencies arrive at different times → measurable with TS-DFT')
```
