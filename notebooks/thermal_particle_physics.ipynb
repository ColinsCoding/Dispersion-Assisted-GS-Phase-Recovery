```python
from sympy import *
from IPython.display import display, Markdown
init_printing(use_latex='mathjax')
def s(label, expr=None):
    display(Markdown(f'**{label}**'))
    if expr is not None: display(expr)
def sec(t):  display(Markdown(f'---\n# {t}'))
def sub(t):  display(Markdown(f'## {t}'))
```

# §1 — 3D Displacement Vector

```python
sec('3D Displacement Vector')
x,y,z = symbols('x y z')
x0,y0,z0 = symbols('x_0 y_0 z_0')

r = Matrix([x-x0, y-y0, z-z0])
s('r = (x−x₀)x̂ + (y−y₀)ŷ + (z−z₀)ẑ', r)
mag = sqrt((x-x0)**2 + (y-y0)**2 + (z-z0)**2)
s('|r| =', mag)
r_hat = r / mag
s('r̂ = r/|r| =', r_hat)

sub('This vector is the argument of EVERY field in physics')
k_c, q, eps0 = symbols('k q epsilon_0', positive=True)
s('Coulomb potential:    V(r) = kq/|r|', k_c*q/mag)
s('Gravitational:        Φ(r) = −GM/|r|', -symbols('G M',positive=True)[0]*symbols('G M',positive=True)[1]/mag)
s('Green\'s function:    G(r,r\') = 1/(4π|r−r\'|)')
s('Wave:                 ψ(r,t) = A·exp(i(k·r − ωt))')

sub('Gradient in 3D — how fields change with r')
f = Function('f')(x,y,z)
grad = Matrix([diff(f,x), diff(f,y), diff(f,z)])
s('∇f =', grad)
s('∇(1/|r|) =  (points toward source):')
inv_r = 1/mag
g = Matrix([diff(inv_r,x), diff(inv_r,y), diff(inv_r,z)])
s('', simplify(g))
s('= −r̂/|r|²  ← Coulomb field direction')
```

# §2 — Thermal Physics

```python
sec('Thermal Physics — Calculus Results')
T, V_s, P, N, n, R, k_B = symbols('T V P N n R k_B', positive=True)
U, S, G, F, H_th = symbols('U S G F H', real=True)

sub('Ideal gas — three equivalent forms')
s('PV = nRT      (n = moles, R = 8.314 J/mol·K)')
s('PV = Nk_BT    (N = molecules, k_B = 1.38×10⁻²³ J/K)')
s('P = ρRT/M     (ρ = density, M = molar mass)')

sub('First Law: dU = δQ − δW')
s('dU = δQ − P dV        (work done BY system = P dV)')
s('δQ = dU + P dV')
s('At constant V:  Q = ΔU = nCᵥΔT')
s('At constant P:  Q = nCₚΔT,   Cₚ = Cᵥ + R')
s('Ratio γ = Cₚ/Cᵥ:  monatomic=5/3, diatomic=7/5')

sub('Second Law: dS ≥ δQ/T')
s('Entropy S — measure of disorder')
s('dS = δQ_rev/T          (reversible process)')
s('ΔS = nCᵥ ln(T₂/T₁) + nR ln(V₂/V₁)  (ideal gas)')
T1,T2,V1,V2 = symbols('T_1 T_2 V_1 V_2', positive=True)
Cv = symbols('C_v', positive=True)
dS = n*Cv*log(T2/T1) + n*R*log(V2/V1)
s('ΔS =', dS)
s('Boltzmann: S = k_B ln(Ω)  (Ω = number of microstates)')

sub('Thermodynamic potentials — which to use when')
potentials = [
    ('Internal energy U',  'S, V constant', 'dU = T dS − P dV'),
    ('Enthalpy H=U+PV',   'S, P constant', 'dH = T dS + V dP'),
    ('Helmholtz F=U−TS',  'T, V constant', 'dF = −S dT − P dV'),
    ('Gibbs G=H−TS',      'T, P constant', 'dG = −S dT + V dP'),
]
for name, natural_vars, differential in potentials:
    s(f'{name:<25} natural vars: {natural_vars:<20} {differential}')

sub('Maxwell relations — from mixed partial derivatives')
s('∂²F/∂T∂V = ∂²F/∂V∂T  →  (∂S/∂V)_T = (∂P/∂T)_V')
s('∂²G/∂T∂P = ∂²G/∂P∂T  →  (∂S/∂P)_T = −(∂V/∂T)_P')
s('These connect unmeasurable quantities to measurable ones')

sub('Heat capacity from calculus')
s('Cᵥ = (∂U/∂T)_V')
s('Cₚ = (∂H/∂T)_P')
s('Cₚ − Cᵥ = −T(∂P/∂V)_T · [(∂V/∂T)_P]²')
s('For ideal gas: Cₚ − Cᵥ = R  (exact)')

sub('Adiabatic process — no heat exchange')
gamma = symbols('gamma', positive=True)
s('δQ = 0  →  dU = −P dV')
s('TV^(γ−1) = const')
s('PV^γ = const')
s('T·P^(−γ/(γ−1)) = const')
s('Speed of sound: v = √(γP/ρ) = √(γRT/M)')
rho, M_mol = symbols('rho M', positive=True)
v_sound = sqrt(gamma*R*T/M_mol)
s('v_sound =', v_sound)
s('At T=293K, air (γ=7/5, M=0.029): v ≈ 343 m/s')
```

# §3 — Statistical Mechanics

```python
sec('Statistical Mechanics — Probability → Thermodynamics')
T, k_B, E_s, Z = symbols('T k_B E Z', positive=True)
beta = 1/(k_B*T)

sub('Boltzmann distribution')
s('P(state s) = exp(−E_s/k_BT) / Z')
s('Z = Σₛ exp(−E_s/k_BT)  (partition function)')
s('β = 1/k_BT  (inverse temperature)')
P_s = exp(-E_s*beta) / Z
s('P(s) =', P_s)

sub('Everything from Z')
s('⟨E⟩ = −∂ln Z/∂β          (mean energy)')
s('F = −k_BT ln Z            (Helmholtz free energy)')
s('S = k_B(ln Z + β⟨E⟩)     (entropy)')
s('Cᵥ = k_Bβ²⟨(ΔE)²⟩        (heat capacity = energy fluctuations)')

sub('Harmonic oscillator — quantum vs classical')
n_s = symbols('n', nonneg=True, integer=True)
hbar, omega_s = symbols('hbar omega', positive=True)
E_n = hbar*omega_s*(n_s + Rational(1,2))
s('Energy levels: Eₙ = ħω(n+½)', E_n)
s('Partition function: Z = Σₙ exp(−βEₙ)')
s('= exp(−βħω/2) · Σₙ exp(−nβħω)')
s('= exp(−βħω/2) / (1 − exp(−βħω))  (geometric series)')
x_s = symbols('x', positive=True)
Z_qho = exp(-x_s/2) / (1 - exp(-x_s))
s('Z (with x=βħω) =', Z_qho)
s('⟨E⟩ = ħω(1/2 + 1/(exp(βħω)−1))')
s('High T limit (βħω<<1): ⟨E⟩ → k_BT  (equipartition)')
s('Low T limit  (βħω>>1): ⟨E⟩ → ħω/2  (zero point energy)')

sub('Maxwell-Boltzmann speed distribution')
v, m_s = symbols('v m', positive=True)
f_MB = 4*pi*n*(m_s/(2*pi*k_B*T))**Rational(3,2) * v**2 * exp(-m_s*v**2/(2*k_B*T))
s('f(v) = 4πn(m/2πk_BT)^(3/2) · v² · exp(−mv²/2k_BT)')
s('Most probable speed v_p = √(2k_BT/m)')
s('Mean speed        ⟨v⟩ = √(8k_BT/πm)')
s('RMS speed         v_rms = √(3k_BT/m)')
v_p   = sqrt(2*k_B*T/m_s)
v_rms = sqrt(3*k_B*T/m_s)
s('v_p =', v_p)
s('v_rms =', v_rms)
s('Ratio v_p : ⟨v⟩ : v_rms = 1 : 1.128 : 1.225')
```

# §4 — Particle Physics

```python
sec('Particle Physics — Standard Model Mathematics')
E, p, m, c, hbar = symbols('E p m c hbar', positive=True)

sub('Special relativity — foundation')
s('Energy-momentum relation (exact):')
E_rel = sqrt((p*c)**2 + (m*c**2)**2)
s('E² = (pc)² + (mc²)²  →  E =', E_rel)
s('Rest energy: E₀ = mc²')
s('Kinetic energy: K = E − mc² = mc²(γ−1)')
beta_s, gamma_s = symbols('beta gamma', positive=True)
s('Lorentz factor: γ = 1/√(1−β²),  β = v/c')
s('Relativistic momentum: p = γmv')
s('Massless particles (photon): E = pc,  v = c always')

sub('Natural units — particle physicists set ħ=c=1')
s('Then: [energy] = [mass] = [momentum] = [1/length] = [1/time]')
s('Electron mass: mₑ = 0.511 MeV')
s('Proton mass:   mₚ = 938.3 MeV')
s('W boson mass:  mW = 80.4 GeV')
s('Higgs mass:    mH = 125.1 GeV')
s('Planck mass:   mₚₗ = 1.22×10¹⁹ GeV  (gravity scale)')

sub('Standard Model — particle table')
particles = [
    ('QUARKS',      '', ''),
    ('up (u)',      '2/3 e',   '2.2 MeV'),
    ('down (d)',    '−1/3 e',  '4.7 MeV'),
    ('charm (c)',   '2/3 e',   '1.28 GeV'),
    ('strange (s)', '−1/3 e', '96 MeV'),
    ('top (t)',     '2/3 e',   '173 GeV'),
    ('bottom (b)', '−1/3 e',  '4.18 GeV'),
    ('LEPTONS',    '', ''),
    ('electron e', '−e',       '0.511 MeV'),
    ('muon μ',     '−e',       '105.7 MeV'),
    ('tau τ',      '−e',       '1.777 GeV'),
    ('νₑ νμ ντ',   '0',        '< 0.1 eV'),
    ('BOSONS',     '', ''),
    ('photon γ',   '0',        '0'),
    ('W±',         '±e',       '80.4 GeV'),
    ('Z⁰',         '0',        '91.2 GeV'),
    ('gluon g',    '0',        '0  (8 types)'),
    ('Higgs H',    '0',        '125.1 GeV'),
]
s(f'{"Particle":<20} {"Charge":<15} {"Mass"}')
for name, charge, mass in particles:
    if charge == '':
        s(f'── {name} ──')
    else:
        s(f'{name:<20} {charge:<15} {mass}')

sub('Feynman diagram → matrix element → cross section')
alpha_em = symbols('alpha', positive=True)
s('Fine structure constant α = e²/4πε₀ħc ≈ 1/137')
s('Each vertex in Feynman diagram: factor of √α ≈ 0.085')
s('Amplitude M ~ α^(n_vertices/2)')
s('Cross section σ ~ |M|² ~ α^n_vertices')
s('e⁺e⁻→μ⁺μ⁻: 2 vertices → σ ~ α²  (tree level)')
s('QED correction: next order ~ α³, then α⁴, ...')
s('Series converges: α << 1  (QED is perturbative)')
s('QCD strong coupling αs ~ 1 at low energy → non-perturbative!')

sub('Decay width and lifetime')
Gamma, tau_p = symbols('Gamma tau', positive=True)
s('Γ = 1/τ  (decay width = inverse lifetime)')
s('N(t) = N₀ · exp(−Γt) = N₀ · exp(−t/τ)')
s('Branching ratio BR = Γᵢ/Γ_total')
lifetimes = [
    ('Muon μ',     '2.2 μs',    'weak decay'),
    ('Neutron n',  '880 s',     'weak decay'),
    ('Pion π⁰',   '8.5×10⁻¹⁷ s', 'EM decay → γγ'),
    ('Top quark t','5×10⁻²⁵ s', 'weak, decays before hadronizing'),
    ('Higgs H',   '1.6×10⁻²² s', 'many channels'),
]
for particle, lifetime, mode in lifetimes:
    s(f'{particle:<15} τ = {lifetime:<20} ({mode})')
```
