# Minimal Coupling → Dispersion: One Mathematical Structure

The substitution $p_\mu \to p_\mu - eA_\mu$ in QED and $k \to k + \Delta k(\omega)$ in dispersive optics
are the same operation: **momentum shifted by a background field**.
This notebook derives both from first principles and makes the analogy exact.

```python
from sympy import *
from IPython.display import display, Markdown

init_printing(use_latex='mathjax')

def step(label, expr=None):
    display(Markdown(f'**{label}**'))
    if expr is not None:
        display(expr)

def section(title):
    display(Markdown(f'---\n## {title}'))
```

## §1 — Free Particle: Dispersion Relation Without a Field

```python
section('Free dispersion relations')

hbar, m, c_light, omega, k = symbols('hbar m c omega k', positive=True)

# Quantum free particle
E_qm = hbar**2 * k**2 / (2*m)
step('Quantum free particle: E = p²/2m = ħ²k²/2m =', E_qm)
step('Dispersion: ω = ħk²/2m  →  quadratic, dispersive')

# Photon in vacuum
E_photon = hbar * c_light * k
step('Photon in vacuum: E = ħck =', E_photon)
step('Dispersion: ω = ck  →  linear, non-dispersive')

# Photon in medium
n_omega = Function('n')(omega)  # frequency-dependent refractive index
display(Markdown(r'''
**Photon in dispersive medium**: $k(\omega) = \frac{\omega\, n(\omega)}{c}$

Taylor expand $k(\omega)$ around carrier $\omega_0$:
$$k(\omega) = k_0 + k_1(\omega-\omega_0) + \frac{1}{2}k_2(\omega-\omega_0)^2 + \cdots$$
where $k_1 = 1/v_g$ (inverse group velocity), $k_2 = \beta_2$ (GVD).
'''))

omega0, k0, k1, k2 = symbols('omega_0 k_0 k_1 k_2', real=True)
dw = symbols('delta_omega', real=True)
k_taylor = k0 + k1*dw + Rational(1,2)*k2*dw**2
step('k(ω₀+δω) ≈', k_taylor)
```

## §2 — Minimal Coupling: $p_\mu \to p_\mu - eA_\mu$

```python
section('Minimal coupling in QED')

display(Markdown(r'''
**Free Schrödinger equation** (operator form):
$$i\hbar\frac{\partial\psi}{\partial t} = \frac{\hat{p}^2}{2m}\psi = -\frac{\hbar^2}{2m}\nabla^2\psi$$

**Demand local U(1) gauge invariance**: $\psi \to e^{i\alpha(\mathbf{r},t)}\psi$

The equation is NOT invariant — derivatives of $\alpha$ survive.
Fix: replace every derivative with the **covariant derivative**:
$$\partial_\mu \to D_\mu = \partial_\mu + \frac{ie}{\hbar}A_\mu$$

Equivalently, replace momentum:
$$\boxed{p_\mu \to p_\mu - eA_\mu}$$

where $A_\mu = (V/c,\ -\mathbf{A})$ is the 4-vector potential.
'''))

# Hamiltonian with minimal coupling
e_charge, V_pot = symbols('e V', real=True)
p_vec = Matrix(symbols('p_x p_y p_z', real=True))
A_vec = Matrix(symbols('A_x A_y A_z', real=True))

# H = (p - eA)²/2m + eV
p_minus_eA = p_vec - e_charge*A_vec
H_coupled = p_minus_eA.dot(p_minus_eA) / (2*m) + e_charge*V_pot
step('H = (p−eA)²/2m + eV =', expand(H_coupled))
step('Expanding: p²/2m − (e/2m)(p·A + A·p) + e²A²/2m + eV')
step('The cross term  −(e/m)A·p  is the interaction: field shifts momentum')
```

## §3 — Same Structure in Dispersive Optics

```python
section('Dispersion as momentum shift')

display(Markdown(r'''
An optical pulse in a dispersive fiber obeys the **nonlinear Schrödinger equation**
(NLSE, paraxial approximation):

$$i\frac{\partial A}{\partial z} = -\frac{\beta_2}{2}\frac{\partial^2 A}{\partial t^2} + \gamma|A|^2 A$$

In frequency domain, the dispersive term acts on the spectral field $\tilde{A}(\omega)$:
$$\tilde{A}(z,\omega) = \tilde{A}(0,\omega)\cdot e^{i\frac{\beta_2}{2}\omega^2 z}$$

The phase accumulated is $\phi(\omega) = \frac{\beta_2 z}{2}\omega^2$.

The **wavenumber is shifted**:
$$k(\omega) = k_0(\omega) + \Delta k(\omega), \quad \Delta k = \frac{\beta_2}{2}\omega^2$$
'''))

beta2, z_fiber, omega_sym = symbols('beta_2 z omega', real=True)

# Dispersion phase
phi_disp = Rational(1,2)*beta2*omega_sym**2*z_fiber
step('Dispersion phase φ(ω,z) =', phi_disp)

# k-shift
Delta_k = diff(phi_disp, z_fiber)
step('Δk = dφ/dz =', Delta_k)

# Group delay = derivative of phase wrt omega
tau_g = diff(phi_disp, omega_sym)
step('Group delay τ_g = dφ/dω =', tau_g)
step('= β₂zω  →  each frequency arrives at a different time (TS-DFT)')

display(Markdown(r'''
**The analogy made explicit**:

| QED (matter in EM field) | Dispersive optics |
|---|---|
| momentum $p$ | wavenumber $k$ |
| gauge field $eA_\mu$ | dispersion $\Delta k(\omega) = \frac{\beta_2}{2}\omega^2$ |
| minimal coupling $p \to p - eA$ | dispersion shift $k \to k + \Delta k$ |
| covariant derivative $D_\mu = \partial_\mu + ieA_\mu/\hbar$ | $\partial_z \to \partial_z + i\frac{\beta_2}{2}\partial_t^2$ |
| free propagator $e^{ip^2 z/2\hbar m}$ | dispersive propagator $e^{i\beta_2\omega^2 z/2}$ |
| phase from A along path: $e\int A_\mu dx^\mu$ | phase from dispersion: $\frac{\beta_2 z}{2}\omega^2$ |

Both are **quadratic phase accumulation** of a wave as it propagates through a background field.
The field is $A_\mu$ in QED; the field is $\beta_2(z)$ (the GVD profile) in fiber optics.
'''))
```

## §4 — Aharonov-Bohm: Phase from Path Integral of A

```python
section('Aharonov-Bohm effect')

display(Markdown(r'''
**Aharonov-Bohm**: a charged particle acquires a phase from the vector potential
even where $\mathbf{B}=0$:
$$\Delta\phi_{AB} = \frac{e}{\hbar}\oint \mathbf{A}\cdot d\mathbf{l} = \frac{e\Phi_B}{\hbar}$$

The particle never touches the field — the potential alone shifts the phase.
This is measurable as interference fringe shift.

**Optical analog** (your experiment):
The optical field never "touches" the GVD profile in any local sense,
but the phase accumulated along the fiber path is:
$$\phi = \int_0^L \Delta k(\omega, z)\,dz = \frac{\beta_2 L}{2}\omega^2$$

Two paths through different dispersions → different phases → interference → measurable.
TDGSA recovers this phase from intensity only — same as AB but for photons.
'''))

# Phase difference between two dispersions
beta2_1, beta2_2, L_sym = symbols('beta_{2,1} beta_{2,2} L', real=True)
phi1 = Rational(1,2)*beta2_1*omega_sym**2*L_sym
phi2 = Rational(1,2)*beta2_2*omega_sym**2*L_sym
delta_phi = phi1 - phi2
step('Phase difference Δφ = φ₁ − φ₂ =', delta_phi)
step('TDGSA measures |e^{iφ₁}|² and |e^{iφ₂}|² separately, recovers Δφ via GS')
```

## §5 — ADC: Where the Continuum Becomes Finite

```python
section('ADC: sampling the phase')

display(Markdown(r'''
The continuous optical field $A(t) \in \mathbb{C}$ gets mapped to:
$$A(t) \xrightarrow{\text{photodetector}} I(t) = |A(t)|^2 \in \mathbb{R}^+
\xrightarrow{\text{ADC}} I[n] \in \mathbb{Z}_{2^N}$$

Three losses in that chain:
1. **Phase** — photodetector squares, phase is gone (the phase retrieval problem)
2. **Continuity** — ADC quantizes to $2^N$ levels ($N=8$: 256 levels)
3. **Bandwidth** — sampled at $f_s$, Nyquist: only $f < f_s/2$ survives
'''))

N_bits, f_s_sym, V_range = symbols('N f_s V_range', positive=True)

# Quantization step
delta_V = V_range / 2**N_bits
step('Quantization step ΔV = V_range / 2^N =', delta_V)

# Quantization noise (uniform distribution)
sigma_q = delta_V / sqrt(12)
step('RMS quantization noise σ_q = ΔV/√12 =', sigma_q)

# SQNR
SQNR_dB = 20*log(2**N_bits * sqrt(6), 10)
step('SQNR = 6.02N + 1.76 dB, at N bits:', simplify(SQNR_dB))
for n in [8, 10, 12, 14]:
    print(f'  {n}-bit ADC: SQNR = {6.02*n + 1.76:.1f} dB')

display(Markdown(r'''
**Connection to phase retrieval**:
Quantization noise $\sigma_q$ propagates into the recovered phase via error propagation:
$$\sigma_\phi \approx \frac{\sigma_q}{|A|} = \frac{\Delta V}{\sqrt{12}\cdot|A|}$$

Phase accuracy improves with:
- More ADC bits (lower $\sigma_q$)
- Higher signal amplitude $|A|$ (more photons)
- More GS iterations (algorithm converges toward Cramér-Rao bound)
- Larger $|\beta_2 z|$ (more diversity between the two measurements)
'''))
```

## §6 — The Theory: Dispersive Phase Retrieval as Inverse Problem

```python
section('TDGSA as structured inverse problem')

display(Markdown(r'''
**What you measure**:
$$I_t[n] = |\psi(t_n)|^2, \quad I_d[n] = |\mathcal{D}\{\psi\}(t_n)|^2$$

where $\mathcal{D}$ is the dispersion operator:
$$\widetilde{\mathcal{D}\{\psi\}}(\omega) = \tilde{\psi}(\omega)\cdot e^{i\frac{\beta_2 z}{2}\omega^2}$$

**What you want**: the complex field $\psi(t) = |\psi(t)|e^{i\phi(t)}$

**Why it works** (the physics reason):
The dispersion operator $\mathcal{D}$ is a **unitary transformation** on $L^2$.
It rotates the field in the space of square-integrable functions.
Two measurements from two different rotations of the same vector
over-constrain the phase — same as triangulation.

**Formal statement**:
$$\phi^* = \arg\min_\phi \|\, |\mathcal{F}^{-1}[\sqrt{I_d}\cdot e^{i\phi_d}]|^2 - I_t \,\|_2^2$$
subject to: $\phi_d(\omega) = \phi(\omega) + \frac{\beta_2 z}{2}\omega^2$

GS is a **coordinate descent** on this objective — alternately projecting onto
the two constraint sets (time intensity, dispersed intensity).
'''))

# Show the constraint sets geometrically
display(Markdown(r'''
**Geometry of GS** (why it converges):

Define two constraint sets in the space of complex fields:
$$\mathcal{C}_t = \{\psi : |\psi(t)|^2 = I_t(t)\} \quad \text{(time intensity constraint)}$$
$$\mathcal{C}_d = \{\psi : |\mathcal{D}\psi(t)|^2 = I_d(t)\} \quad \text{(dispersive constraint)}$$

Both are **non-convex** manifolds in $\mathbb{C}^N$.
GS = alternating projections onto $\mathcal{C}_t$ and $\mathcal{C}_d$.
Converges when the manifolds intersect transversally — guaranteed when
$|\beta_2 z| \geq 5000$ ps² (sufficient diversity, correlation $< 0.95$).
'''))
```
