# Complex Numbers · Polar Form · Logarithm
Real-imaginary plane → Euler's formula → phasors → signal phase → defense sensing.

````python
from sympy import *
from IPython.display import display, Markdown

init_printing(use_latex='mathjax')

def step(label, expr=None):
    display(Markdown(f'**{label}**'))
    if expr is not None:
        display(expr)

def section(title):
    display(Markdown(f'---\n## {title}'))
````

## §1 — What i Is: Solving x² = −1

````python
section('Imaginary unit')

x = symbols('x')

step('Real equation with no real solution: x² + 1 = 0')
step('Real solutions:', solve(x**2 + 1, x))
step('→ Define i = √(−1), so i² = −1')

# Powers of i
display(Markdown(r'''
**Powers of i cycle with period 4**:
$$i^0=1,\quad i^1=i,\quad i^2=-1,\quad i^3=-i,\quad i^4=1,\ldots$$
'''))
for n in range(8):
    print(f'  i^{n} = {I**n}')

# Complex number
a, b = symbols('a b', real=True)
z = a + I*b
step('\nComplex number z = a + ib:', z)
step('Real part Re(z) =', re(z))
step('Imaginary part Im(z) =', im(z))
step('Complex conjugate z* =', conjugate(z))
step('Modulus |z| = √(a²+b²) =', Abs(z))
step('|z|² = z·z* =', expand(z * conjugate(z)))
````

## §2 — The Complex Plane: Cartesian and Polar

````python
section('Polar form')

r, theta = symbols('r theta', real=True)

display(Markdown(r'''
Every complex number lives on a 2D plane:
- Horizontal axis: real part
- Vertical axis: imaginary part

**Cartesian**: $z = a + ib$  
**Polar**: $z = r e^{i\theta}$

Conversion:
$$r = |z| = \sqrt{a^2+b^2}, \quad \theta = \arg(z) = \arctan(b/a)$$
$$a = r\cos\theta, \quad b = r\sin\theta$$
'''))

# Euler's formula
step('Euler formula: e^(iθ) =', exp(I*theta).rewrite(cos))
step('= cos(θ) + i·sin(θ)')

# Verify by Taylor series
step('\nVerify by Taylor series of e^(iθ):')
s = series(exp(I*theta), theta, 0, 8)
step('e^(iθ) =', s)
step('Real terms = cos(θ), imaginary terms = i·sin(θ)  ✓')

# Most beautiful equation
step("Euler's identity (θ=π):", Eq(exp(I*pi) + 1, 0))
step('e^(iπ) + 1 = 0  — connects e, i, π, 1, 0')
````

## §3 — Complex Arithmetic in Polar Form

````python
section('Arithmetic')

r1, r2, t1, t2 = symbols('r_1 r_2 theta_1 theta_2', real=True, positive=True)
z1 = r1 * exp(I*t1)
z2 = r2 * exp(I*t2)

# Multiplication: add angles
prod = z1 * z2
step('z₁·z₂ =', prod)
step('= r₁r₂ · e^(i(θ₁+θ₂))  →  multiply magnitudes, ADD angles')

# Division: subtract angles
div = z1 / z2
step('z₁/z₂ =', simplify(div))
step('= (r₁/r₂) · e^(i(θ₁−θ₂))  →  divide magnitudes, SUBTRACT angles')

# Powers: De Moivre
n_sym = symbols('n', integer=True)
step('De Moivre: zⁿ = rⁿ·e^(inθ) = rⁿ(cos(nθ) + i·sin(nθ))')

# Roots
display(Markdown(r'''
**nth roots of unity**: solutions to $z^n = 1$
$$z_k = e^{2\pi i k/n}, \quad k = 0, 1, \ldots, n-1$$
These are equally spaced on the unit circle — the **DFT frequencies**.
'''))
for k in range(4):
    zk = exp(2*pi*I*k/4)
    step(f'4th root k={k}: z =', simplify(zk))
````

## §4 — Logarithm: Real, Then Complex

````python
section('Logarithm')

x_pos = symbols('x', positive=True)

# Real log
step('Natural log: ln(x) = ∫₁ˣ (1/t)dt')
step('Inverse of exp: ln(eˣ) =', simplify(log(exp(x_pos))))
step('Product rule: ln(xy) =', expand_log(log(x_pos * symbols('y', positive=True))))
step('Power rule: ln(xⁿ) =', expand_log(log(x_pos**symbols('n'))))

# Change of base
step('Change of base: log_b(x) = ln(x)/ln(b)')
step('log₁₀(x) = ln(x)/ln(10) ≈ ln(x)/2.3026')
step('log₂(x) = ln(x)/ln(2)  — bits')

# Complex log — multi-valued
display(Markdown(r'''
**Complex logarithm** — multi-valued because $e^{i(\theta+2\pi k)} = e^{i\theta}$:
$$\ln(re^{i\theta}) = \ln(r) + i\theta + 2\pi i k, \quad k \in \mathbb{Z}$$

**Principal value** (branch cut on negative real axis, $\theta \in (-\pi, \pi]$):
$$\text{Log}(z) = \ln|z| + i\,\text{Arg}(z)$$
'''))

z_ex = -1  # on branch cut
step('Log(−1) =', log(S(-1)))
step('= iπ  (Euler: e^(iπ) = −1 → ln(−1) = iπ)')
step('Log(i) =', log(I))
step('Log(1+i) =', log(1+I))
step('= ln(√2) + iπ/4 =', simplify(log(1+I)))
````

## §5 — Polar Form in Physics: Phasors and AC Circuits

````python
section('Phasors')

omega, t, phi_0, V0 = symbols('omega t phi V_0', real=True)

# Real signal
V_real = V0 * cos(omega*t + phi_0)
step('AC voltage: V(t) =', V_real)
step('= Re[V₀·e^(i(ωt+φ))]  →  phasor representation')

# Phasor
V_phasor = V0 * exp(I*phi_0)  # drop e^(iωt), implicit
step('Phasor Ṽ = V₀·e^(iφ) =', V_phasor)
step('Magnitude = amplitude, angle = phase')

# Impedance
R, L, C = symbols('R L C', positive=True)
Z_R = R
Z_L = I*omega*L
Z_C = 1/(I*omega*C)
step('\nImpedance:')
step('Z_R =', Z_R, '  (real → in phase)')
step('Z_L =', Z_L, '  (imaginary → voltage leads current by 90°)')
step('Z_C =', Z_C, '  (imaginary → current leads voltage by 90°)')

# Series RLC
Z_RLC = Z_R + Z_L + Z_C
step('Series RLC: Z =', Z_RLC)
step('|Z|² =', simplify(Z_RLC * conjugate(Z_RLC)))

# Resonance
omega_res = 1/sqrt(L*C)
step('Resonance (Im[Z]=0): ω₀ =', omega_res)
step('At resonance: Z = R  →  purely resistive, maximum current')
````

## §6 — dB Scale: Logarithm in Engineering

````python
section('Decibels')

P1, P2 = symbols('P_1 P_2', positive=True)

dB_power = 10*log(P2/P1, 10)
step('Power ratio in dB: 10·log₁₀(P₂/P₁) =', dB_power)

dB_amplitude = 20*log(symbols('A_2 A_1', positive=True)[0]/symbols('A_2 A_1', positive=True)[1], 10)
step('Amplitude ratio: 20·log₁₀(A₂/A₁)  (because P∝A²)')

display(Markdown(r'''
| Ratio | Power dB | What it means |
|---|---|---|
| 2× | +3 dB | double power |
| 10× | +10 dB | decade |
| 100× | +20 dB | two decades |
| 0.5× | −3 dB | half power (3dB bandwidth) |
| 0.001× | −30 dB | sidelobe level in phased array |
| 10⁻⁹× | −90 dB | isolation in fiber tap |
'''))

# Bode plot: log frequency axis
display(Markdown(r'''
**Bode plot**: $|H(i\omega)|$ in dB vs $\log_{10}(\omega)$

First-order low-pass $H = 1/(1+i\omega\tau)$:
$$|H|_{dB} = -10\log_{10}(1+(\omega\tau)^2)$$
- Below cutoff $\omega \ll 1/\tau$: 0 dB (flat)
- Above cutoff $\omega \gg 1/\tau$: −20 dB/decade slope
'''))

tau, omega_sym = symbols('tau omega', positive=True)
H_lp = 1/(1 + I*omega_sym*tau)
H_mag_dB = 10*log(H_lp*conjugate(H_lp), 10)
step('|H|² in dB =', simplify(H_mag_dB))
step('At ω=1/τ (cutoff):', simplify(H_mag_dB.subs(omega_sym, 1/tau)))
````

## §7 — Phase in Signal Processing: FFT and TDGSA

````python
section('Phase in the DFT')

display(Markdown(r'''
The DFT of a signal $x[n]$ is:
$$X[k] = \sum_{n=0}^{N-1} x[n]\, e^{-2\pi i kn/N} = |X[k]|\, e^{i\phi_k}$$

$|X[k]|$ = amplitude spectrum  
$\phi_k = \arg(X[k]) = \text{Im}(\ln X[k])$ = **phase spectrum**

The phase spectrum encodes **timing and structure** of the signal.
The amplitude spectrum encodes **which frequencies** are present.
'''))

import numpy as np
N = 64
n = np.arange(N)

# Gaussian pulse at center
x = np.exp(-((n-N//2)/5)**2)
X = np.fft.fft(x)
amp = np.abs(X)
phase = np.angle(X)  # = Im(log(X))

print(f'Gaussian pulse: peak amplitude at k={np.argmax(amp)}')
print(f'Phase at k=0: {phase[0]:.4f} rad')
print(f'Phase at k=1: {phase[1]:.4f} rad')
print(f'Phase slope (group delay): {np.diff(np.unwrap(phase[:10]))[:3]}  rad/bin')

display(Markdown(r'''
**TDGSA recovers $\phi_k$** from two intensity measurements:
$$|X[k]|^2 = I_{\text{freq}}[k] \quad \text{(known)}$$
$$|\mathcal{D}\{x\}[k]|^2 = I_{\text{disp}}[k] \quad \text{(known)}$$
$$\phi_k = ? \quad \text{(unknown → GS recovers it)}$$

The logarithm connection:
$$\phi_k = \text{Im}(\ln X[k]) = \text{Im}(\ln|X[k]| + i\phi_k) = \text{Im}(\text{Log}(X[k]))$$

Phase retrieval = recovering the imaginary part of the complex logarithm of the signal spectrum.
'''))
````

## §8 — Branch Cuts and Phase Unwrapping

````python
section('Branch cuts in practice')

display(Markdown(r'''
The complex logarithm has a **branch cut** — $\arg(z)$ jumps by $2\pi$
when you cross the negative real axis.

In discrete signals: $\phi[k] \in (-\pi, \pi]$ (wrapped).
When the true phase exceeds $\pi$, it wraps around — looks like a jump.

**Phase unwrapping**: add multiples of $2\pi$ to remove jumps:
$$\phi_{\text{unwrapped}}[k] = \phi[k] + 2\pi m_k$$
where $m_k$ is chosen so $|\phi[k] - \phi[k-1]| < \pi$.

This is critical for TDGSA: the recovered phase must be unwrapped
before computing group delay $\tau_g = d\phi/d\omega$.
'''))

import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

# Chirp signal: linearly increasing phase
N = 128
k = np.arange(N)
true_phase = 0.15 * k**2 / N  # quadratic phase (like dispersion)
wrapped = np.angle(np.exp(1j * true_phase))
unwrapped = np.unwrap(wrapped)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,3))
ax1.plot(k, wrapped)
ax1.set_title('Wrapped phase ∈ (−π, π]')
ax1.set_xlabel('k')
ax1.set_ylabel('φ (rad)')
ax2.plot(k, unwrapped, color='orange')
ax2.plot(k, true_phase, 'k--', label='true')
ax2.set_title('Unwrapped = true quadratic phase')
ax2.set_xlabel('k')
ax2.legend()
plt.tight_layout()
plt.savefig('phase_unwrap_demo.png', dpi=100)
plt.show()

print(f'Max error after unwrap: {np.max(np.abs(unwrapped - true_phase)):.2e} rad')
````
