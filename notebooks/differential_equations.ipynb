# Differential Equations — Complete
### First Order → Second Order → Systems → PDEs
Every step. No gaps.

```python
from sympy import *
from IPython.display import display, Markdown
init_printing(use_latex='mathjax')
def s(label, expr=None):
    display(Markdown(f'**{label}**'))
    if expr is not None: display(expr)
def sec(t):  display(Markdown(f'---\n# {t}'))
def sub(t):  display(Markdown(f'## {t}'))
```

# §1 — Classification & Vocabulary

```python
sec('Classification')
x = symbols('x')
y = Function('y')

s('ORDER = highest derivative present')
s('DEGREE = power of highest derivative (when polynomial)')
s('LINEAR = y and all derivatives appear to first power, no products')
s('')
examples = [
    ("y' = 3y",              "1st order, linear, homogeneous"),
    ("y'' + 2y' + y = 0",   "2nd order, linear, homogeneous"),
    ("y'' + 2y' + y = eˣ",  "2nd order, linear, NON-homogeneous"),
    ("y·y' = x",             "1st order, NON-linear"),
    ("(y'')² + y = 0",      "2nd order, NON-linear (degree 2)"),
    ("∂²u/∂t²= c²∂²u/∂x²", "PDE, 2nd order, linear"),
]
for eq, kind in examples:
    s(f'{eq:<35} → {kind}')
```

# §2 — First Order ODEs

```python
sec('First Order ODEs')
x = symbols('x')
y = Function('y')

sub('Separable: dy/dx = g(x)·h(y)  →  dy/h(y) = g(x)dx')
cases = [
    ("dy/dx = y",          Eq(y(x).diff(x), y(x))),
    ("dy/dx = xy",         Eq(y(x).diff(x), x*y(x))),
    ("dy/dx = (1+y²)x",   Eq(y(x).diff(x), (1+y(x)**2)*x)),
]
for label, ode in cases:
    sol = dsolve(ode)
    s(f'ODE: {label}', ode)
    s('  General solution:', sol)
    s('  Verify:', simplify(ode.subs(y(x), sol.rhs).doit()))

sub('Linear 1st order: dy/dx + P(x)y = Q(x)')
s('Integrating factor μ(x) = e^(∫P dx)')
s('Multiply through: d/dx[μy] = μQ')
s('Integrate: μy = ∫μQ dx')

linear_cases = [
    ("y' + y = eˣ",      Eq(y(x).diff(x) + y(x), exp(x))),
    ("y' + (1/x)y = x",  Eq(y(x).diff(x) + y(x)/x, x)),
    ("y' − 2y = 4x",     Eq(y(x).diff(x) - 2*y(x), 4*x)),
]
for label, ode in linear_cases:
    sol = dsolve(ode)
    s(f'ODE: {label}', ode)
    s('  Solution:', sol)

sub('Exact equations: M dx + N dy = 0  where ∂M/∂y = ∂N/∂x')
x, y_s = symbols('x y')
M = 2*x*y_s + 1
N = x**2 + 2*y_s
s('M =', M, )
s('N =', N)
s('∂M/∂y =', diff(M, y_s))
s('∂N/∂x =', diff(N, x))
s('Equal → exact')
F = integrate(M, x) 
s('F = ∫M dx =', F)
g_y = integrate(N - diff(F, y_s), y_s)
s('g(y) from ∂F/∂y = N:', g_y)
s('Solution F(x,y) = C:', F + g_y)
```

# §3 — Second Order Linear ODEs

```python
sec('Second Order Linear ODEs')
x = symbols('x')
y = Function('y')

sub('Homogeneous: ay\'\'+ by\' + cy = 0')
s('Characteristic equation: ar² + br + c = 0')
s('Discriminant Δ = b²− 4ac determines solution form:')

a_s, b_s, c_s, r = symbols('a b c r')
char = a_s*r**2 + b_s*r + c_s
roots = solve(char, r)
s('Roots r₁,r₂ =', roots)

cases_2nd = [
    ('Δ>0 distinct real roots:  y = C₁e^(r₁x) + C₂e^(r₂x)',
     Eq(y(x).diff(x,2) - 5*y(x).diff(x) + 6*y(x), 0)),
    ('Δ=0 repeated root r:      y = (C₁+C₂x)e^(rx)',
     Eq(y(x).diff(x,2) - 4*y(x).diff(x) + 4*y(x), 0)),
    ('Δ<0 complex roots α±iβ:   y = e^(αx)(C₁cos βx + C₂sin βx)',
     Eq(y(x).diff(x,2) - 2*y(x).diff(x) + 5*y(x), 0)),
]
for desc, ode in cases_2nd:
    r_sym = symbols('r')
    coeffs = ode.lhs.as_poly(y(x)).all_coeffs() if False else None
    sol = dsolve(ode)
    s(desc)
    s('  ODE:', ode)
    s('  Solution:', sol)

sub('Non-homogeneous: ay\'\'+ by\' + cy = f(x)')
s('y = y_h + y_p  (homogeneous + particular)')
s('')
s('Method of undetermined coefficients — guess y_p:')
guess_table = [
    ('f(x) = eˣ',         'y_p = Aeˣ'),
    ('f(x) = sin x',      'y_p = A sin x + B cos x'),
    ('f(x) = x²',         'y_p = Ax² + Bx + C'),
    ('f(x) = x·eˣ',       'y_p = (Ax+B)eˣ'),
    ('f(x) = e²ˣsin x',   'y_p = e²ˣ(A sin x + B cos x)'),
]
for f, guess in guess_table:
    s(f'  {f:<25} → guess {guess}')

s('')
nonhom_cases = [
    Eq(y(x).diff(x,2) + y(x), cos(x)),
    Eq(y(x).diff(x,2) - 3*y(x).diff(x) + 2*y(x), exp(x)),
    Eq(y(x).diff(x,2) + 4*y(x), x**2),
]
for ode in nonhom_cases:
    sol = dsolve(ode)
    s('ODE:', ode)
    s('  Solution:', sol)
```

# §4 — Physical Systems as ODEs

```python
sec('Physical Systems')
x_t = Function('x')
t = symbols('t')
m, k, b, omega0, F0, omega = symbols('m k b omega_0 F_0 omega', positive=True)

sub('Simple Harmonic Oscillator: mẍ + kx = 0')
sho = Eq(m*x_t(t).diff(t,2) + k*x_t(t), 0)
s('', sho)
sol_sho = dsolve(sho)
s('Solution:', sol_sho)
s('ω₀ = √(k/m)  (natural frequency):', sqrt(k/m))
s('Period T = 2π/ω₀:', 2*pi/sqrt(k/m))

sub('Damped Oscillator: mẍ + bẋ + kx = 0')
damped = Eq(m*x_t(t).diff(t,2) + b*x_t(t).diff(t) + k*x_t(t), 0)
s('', damped)
s('α = b/2m (damping),  ω₀ = √(k/m) (natural freq)')
alpha = b/(2*m)
for label, b_val, m_val, k_val in [
    ('Underdamped  (α<ω₀)', 1, 1, 10),
    ('Critically damped (α=ω₀)', 2*sqrt(10), 1, 10),
    ('Overdamped   (α>ω₀)', 20, 1, 10),
]:
    ode = Eq(m_val*x_t(t).diff(t,2) + b_val*x_t(t).diff(t) + k_val*x_t(t), 0)
    sol = dsolve(ode)
    s(f'{label}:', sol)

sub('Driven Oscillator: mẍ + bẋ + kx = F₀cos(ωt)')
driven = Eq(x_t(t).diff(t,2) + 2*x_t(t).diff(t) + 10*x_t(t), cos(3*t))
s('Example (m=1, b=2, k=10, F₀cos3t):', driven)
sol_driven = dsolve(driven)
s('Solution:', sol_driven)
s('Steady-state amplitude A(ω) = F₀/√[(k−mω²)² + b²ω²]')
A_omega = F0 / sqrt((k - m*omega**2)**2 + b**2*omega**2)
s('A(ω) =', A_omega)
s('Resonance at ω = ω₀√(1−2α²/ω₀²) ≈ ω₀ for light damping')

sub('RC, RL, RLC circuits (same math)')
R_s, C_s, L_s, V0 = symbols('R C L V_0', positive=True)
q = Function('q')
odes_circuit = [
    ('RC: q\' + q/RC = V₀/R',  Eq(q(t).diff(t) + q(t)/(R_s*C_s), V0/R_s)),
    ('RL: L·I\' + RI = V₀',    Eq(L_s*q(t).diff(t,2) + R_s*q(t).diff(t), V0)),
    ('RLC: Lq\'\' + Rq\' + q/C = V₀', Eq(L_s*q(t).diff(t,2) + R_s*q(t).diff(t) + q(t)/C_s, V0)),
]
for label, ode in odes_circuit:
    s(label)
    sol = dsolve(ode)
    s('  Solution:', sol)
```

# §5 — Systems of ODEs

```python
sec('Systems of ODEs  →  Matrix Exponential')
t = symbols('t')

sub('dx/dt = Ax  →  x(t) = e^(At)x(0)')
s('Solve via eigendecomposition:')
s('  A = PDP⁻¹  →  e^(At) = P·e^(Dt)·P⁻¹')
s('  e^(Dt) = diag(e^(λ₁t), e^(λ₂t), ...)')

A = Matrix([[-1, 2], [0, -3]])
s('Example: A =', A)
P, D = A.diagonalize()
s('Eigenvalues:', list(D.diagonal()))
s('P =', P)
s('D =', D)

lam1, lam2 = D[0,0], D[1,1]
eDt = Matrix([[exp(lam1*t), 0], [0, exp(lam2*t)]])
eAt = P * eDt * P.inv()
s('e^(At) =', simplify(eAt))
s('x(t) = e^(At)·x₀')

sub('Predator-Prey (Lotka-Volterra) — nonlinear system')
s('dx/dt =  αx − βxy   (prey: grows alone, eaten by predator)')
s('dy/dt = −γy + δxy   (predator: dies alone, fed by prey)')
s('Linearize around equilibrium (x*,y*) = (γ/δ, α/β):')
alpha_s, beta_s, gamma_s, delta_s = symbols('alpha beta gamma delta', positive=True)
xstar = gamma_s/delta_s
ystar = alpha_s/beta_s
s('Equilibrium:', (xstar, ystar))
J = Matrix([[ 0,          -beta_s*xstar],
            [delta_s*ystar, 0           ]])
s('Jacobian at equilibrium:', J)
eigs = J.eigenvals()
s('Eigenvalues:', eigs)
s('Pure imaginary → center (neutrally stable oscillations)')

sub('Phase plane analysis')
s('Equilibrium types from eigenvalues λ₁,λ₂:')
phase_table = [
    ('Both real, negative',           'Stable node (sink)'),
    ('Both real, positive',           'Unstable node (source)'),
    ('Real, opposite signs',          'Saddle (unstable)'),
    ('Complex α±iβ, α<0',            'Stable spiral'),
    ('Complex α±iβ, α>0',            'Unstable spiral'),
    ('Pure imaginary ±iβ',           'Center (neutral)'),
]
for eig_type, behavior in phase_table:
    s(f'  {eig_type:<40} → {behavior}')
```

# §6 — Laplace Transform Method

```python
sec('Laplace Transform — ODEs become Algebra')
t, s_var = symbols('t s', positive=True)
y = Function('y')

sub('Key property: ℒ{y\''} = sY(s) − y(0)')
s('ℒ{y\'\''} = s²Y(s) − sy(0) − y\'(0)')
s('ODE in t → algebraic equation in s → solve → inverse transform')

sub('Laplace transform table')
transform_table = [
    ('1',           1/s_var),
    ('t',           1/s_var**2),
    ('tⁿ',          factorial(symbols('n'))/s_var**(symbols('n')+1)),
    ('eᵃᵗ',         1/(s_var-symbols('a'))),
    ('sin(ωt)',      symbols('omega')/(s_var**2+symbols('omega')**2)),
    ('cos(ωt)',      s_var/(s_var**2+symbols('omega')**2)),
    ('e^(at)sin(ωt)', symbols('omega')/((s_var-symbols('a'))**2+symbols('omega')**2)),
    ('δ(t)',        Integer(1)),
    ('u(t)',        1/s_var),
]
for fn, F in transform_table:
    s(f'ℒ{{{fn}}} =', F)

sub('Worked example: y\'\' + 3y\' + 2y = e^(−t), y(0)=0, y\'(0)=0')
s('Apply ℒ: s²Y + 3sY + 2Y = 1/(s+1)')
s('Y(s)(s²+3s+2) = 1/(s+1)')
s('Y(s) = 1/[(s+1)(s²+3s+2)]')
s('     = 1/[(s+1)(s+1)(s+2)]')
s('     = 1/[(s+1)²(s+2)]')
Y_s = 1/((s_var+1)**2 * (s_var+2))
s('Y(s) =', Y_s)
pf = apart(Y_s, s_var)
s('Partial fractions:', pf)
sol_t = inverse_laplace_transform(pf, s_var, t)
s('y(t) = ℒ⁻¹{Y(s)} =', sol_t)

sub('Transfer function H(s) = Y(s)/X(s)')
s('For ay\'\' + by\' + cy = x(t):')
a_c,b_c,c_c = symbols('a b c', positive=True)
H = 1/(a_c*s_var**2 + b_c*s_var + c_c)
s('H(s) = 1/(as²+bs+c) =', H)
s('Poles = roots of denominator = eigenvalues of system')
s('Poles in left half-plane (Re<0) → stable system')
s('Poles on imaginary axis → marginally stable (oscillates forever)')
s('Poles in right half-plane → unstable (blows up)')
```

# §7 — Partial Differential Equations

```python
sec('Partial Differential Equations')
x, t, L = symbols('x t L', positive=True)
c, k_diff, alpha_diff = symbols('c k alpha', positive=True)

sub('The three fundamental PDEs')
u = Function('u')(x,t)
s('Wave equation:  ∂²u/∂t² = c²∂²u/∂x²    (vibrations, EM, acoustics)')
s('Heat equation:  ∂u/∂t  = α∂²u/∂x²      (diffusion, heat, Brownian motion)')
s('Laplace:        ∂²u/∂x² + ∂²u/∂y² = 0  (steady state, electrostatics)')

sub('Heat equation — separation of variables')
s('u(x,t) = X(x)T(t)')
s('T\'/αT = X\'\'/X = −λ  (separation constant)')
s('')
s('Two ODEs:')
X = Function('X')
T_fn = Function('T')
lam = symbols('lambda', positive=True)
ode_X = Eq(X(x).diff(x,2) + lam*X(x), 0)
ode_T = Eq(T_fn(t).diff(t) + alpha_diff*lam*T_fn(t), 0)
s('X\'\'+ λX = 0:', ode_X, )
s('Solution:', dsolve(ode_X))
s('T\' + αλT = 0:', ode_T)
s('Solution:', dsolve(ode_T))
s('')
s('BC: u(0,t)=0, u(L,t)=0  →  X(0)=X(L)=0')
s('Eigenvalues: λₙ = (nπ/L)²')
s('Eigenfunctions: Xₙ = sin(nπx/L)')
n = symbols('n', positive=True, integer=True)
lambda_n = (n*pi/L)**2
X_n = sin(n*pi*x/L)
T_n = exp(-alpha_diff * lambda_n * t)
s('Xₙ =', X_n)
s('Tₙ =', T_n)
s('General solution (Fourier series):')
s('u(x,t) = Σₙ Bₙ sin(nπx/L) e^(−αn²π²t/L²)')
s('Bₙ = (2/L) ∫₀ᴸ u(x,0) sin(nπx/L) dx  (initial condition)')

sub('Wave equation — d\'Alembert + standing waves')
s('d\'Alembert: u(x,t) = f(x−ct) + g(x+ct)')
s('Standing wave on [0,L]: u(x,t) = Σₙ sin(nπx/L)[Aₙcos(nπct/L) + Bₙsin(nπct/L)]')
s('Frequencies: fₙ = nc/2L  (harmonics)')
f_n = n*c/(2*L)
s('fₙ =', f_n)
for n_val in range(1,5):
    s(f'  f_{n_val} = {n_val}·c/2L  (harmonic {n_val})')

sub('Laplace equation — electrostatics')
y_s = symbols('y')
s('∇²u = 0  (harmonic functions)')
s('Solutions are smooth, no local maxima/minima inside domain')
s('Physical: steady-state temperature, electric potential in free space')
s('Separation in 2D rectangle: u(x,y) = X(x)Y(y)')
s('X\'\'/X = −Y\'\'/Y = λ')
s('One oscillates (sin/cos), other decays/grows (sinh/cosh)')
s('BC pins which combination survives')
```
