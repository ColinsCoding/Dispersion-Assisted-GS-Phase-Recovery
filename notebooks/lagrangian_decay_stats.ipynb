```python
from sympy import *
from sympy.vector import CoordSys3D
from IPython.display import display, Markdown
init_printing(use_latex='mathjax')
def s(label, expr=None):
    display(Markdown(f'**{label}**'))
    if expr is not None: display(expr)
def sec(t): display(Markdown(f'---\n# {t}'))
def sub(t): display(Markdown(f'## {t}'))
```

# §1 — Lagrangian Mechanics in Spherical Coordinates

```python
sec('Lagrangian in Spherical Coordinates')
t = symbols('t')
r, theta, phi_s = symbols('r theta phi', positive=True)
m, g = symbols('m g', positive=True)
r_fn     = Function('r')(t)
theta_fn = Function('theta')(t)
phi_fn   = Function('phi')(t)
dr     = r_fn.diff(t)
dtheta = theta_fn.diff(t)
dphi   = phi_fn.diff(t)

sub('Kinetic energy — metric of spherical coords')
s('Cartesian: T = m/2(ẋ²+ẏ²+ż²)')
s('Spherical: x=r sinθ cosφ,  y=r sinθ sinφ,  z=r cosθ')
s('Line element: ds² = dr² + r²dθ² + r²sin²θ dφ²')
T = Rational(1,2)*m*(dr**2 + r_fn**2*dtheta**2 + r_fn**2*sin(theta_fn)**2*dphi**2)
s('T =', T)

sub('Potential and Lagrangian L = T − V')
V = m*g*r_fn*cos(theta_fn)
s('V = mgr cosθ  (gravity, z=r cosθ):', V)
L = T - V
s('L = T − V =', L)

sub('Euler-Lagrange equations: d/dt(∂L/∂q̇) − ∂L/∂q = 0')
coords = [r_fn, theta_fn, phi_fn]
names  = ['r', 'θ', 'φ']
for q, name in zip(coords, names):
    dq = q.diff(t)
    EL = (diff(L, dq).diff(t) - diff(L, q)).simplify()
    s(f'EL equation for {name}:', Eq(EL, 0))

sub('Conserved quantities from symmetry (Noether theorem)')
s('φ is cyclic (L independent of φ):')
p_phi = diff(L, dphi)
s('  p_φ = ∂L/∂φ̇ = mr²sin²θ · φ̇ =', p_phi)
s('  p_φ = const  ← z-angular momentum conserved')
s('If also θ=π/2 (equatorial): p_φ = mr²φ̇ = L_z')
s('Energy E = T + V conserved (no explicit t dependence)')
```

# §2 — Lagrangian in N Dimensions

```python
sec('Lagrangian in N Dimensions (2D, 3D, 4D, 5D)')
t = symbols('t')
m = symbols('m', positive=True)

sub('General N-dim kinetic energy')
s('T = m/2 · Σᵢ gᵢⱼ q̇ᵢ q̇ⱼ  (metric tensor gᵢⱼ)')
s('Flat space: gᵢⱼ = δᵢⱼ  →  T = m/2 Σᵢ q̇ᵢ²')

for N in [2, 3, 4, 5]:
    qs   = [Function(f'q{i+1}')(t) for i in range(N)]
    dqs  = [q.diff(t) for q in qs]
    T_N  = Rational(1,2)*m*sum(dq**2 for dq in dqs)
    s(f'T ({N}D, flat):', T_N)

sub('4D spacetime Lagrangian (special relativity)')
s('Minkowski metric: ds² = −c²dt² + dx² + dy² + dz²')
s('Action: S = −mc ∫ ds = −mc² ∫ √(1−v²/c²) dt')
c_s, v_s = symbols('c v', positive=True)
L_rel = -m*c_s**2 * sqrt(1 - v_s**2/c_s**2)
s('L_relativistic =', L_rel)
s('Expand for v<<c:')
v_small = symbols('epsilon', positive=True)
approx = series(L_rel.subs(v_s, v_small*c_s), v_small, 0, 4)
s('≈', approx)
s('= −mc² + mv²/2 + ...  (rest energy + Newtonian KE)')

sub('5D Kaluza-Klein (gravity + EM unified)')
s('5th dimension compactified on circle of radius R')
s('5D metric: g_AB (A,B = 0,1,2,3,5)')
s('Decompose: g_μν = 4D metric, g_μ5 = EM 4-potential Aμ, g_55 = scalar φ')
s('5D Einstein action → 4D Einstein + Maxwell + scalar')
s('Unifies gravity (gμν) and electromagnetism (Aμ) geometrically')
s('Particle in 5D: 5th momentum p_5 = e/c  (electric charge!)')
s('Charge quantized if 5th dimension is compact')
```

# §3 — Alpha Decay

```python
sec('Alpha Decay — Quantum Tunneling Through Coulomb Barrier')
E, V0, r, a, hbar, m = symbols('E V_0 r a hbar m', positive=True)

sub('What happens')
s('Parent nucleus → daughter + ⁴He (alpha particle, Z=2, A=4)')
s('²³⁸U → ²³⁴Th + ⁴He  +  4.27 MeV')
s('Alpha is trapped by nuclear potential but tunnels through Coulomb barrier')

sub('Q-value — energy released')
s('Q = (M_parent − M_daughter − M_alpha) · c²')
s('Q > 0 → exothermic, alpha flies off with KE ≈ Q')
s('Most of KE goes to alpha (conservation of momentum):')
M_p, M_d, M_a = symbols('M_p M_d M_a', positive=True)
s('KE_alpha = Q · M_daughter/(M_daughter + M_alpha)')

sub('Gamow tunneling — WKB approximation')
s('Barrier: Coulomb repulsion V(r) = 2Ze²/4πε₀r  for r > R_nuclear')
s('Alpha has energy E < V(r) in barrier region → classically forbidden')
s('Quantum: wavefunction decays exponentially through barrier')
s('')
s('Transmission probability:')
s('T ≈ exp(−2G)  where G = Gamow factor')
s('G = (1/ħ) ∫_{R₁}^{R₂} √(2m(V(r)−E)) dr')
Z_d, e_c, eps0_s, r_s = symbols('Z e epsilon_0 r', positive=True)
V_coulomb = 2*Z_d*e_c**2/(4*pi*eps0_s*r_s)
s('V_Coulomb(r) =', V_coulomb)
s('')
s('Gamow integral (analytical result):')
s('G ≈ π·Z·e²/(2ε₀hv_alpha)  − 4√(2mR_c·E)/ħ')
s('where v_alpha = speed of alpha at infinity')

sub('Geiger-Nuttall law')
s('log(λ) = A − B/√E_alpha')
s('λ = decay constant = Γ/ħ = 1/τ')
s('Span 20 orders of magnitude in half-life:')
import math
nuclei = [
    ('²¹²Po', 8.95,  '0.3 μs'),
    ('²¹⁰Po', 5.30,  '138 days'),
    ('²²⁶Ra', 4.87,  '1600 years'),
    ('²³⁸U',  4.27,  '4.5 billion years'),
]
s(f'{"Nucleus":<10} {"E_α (MeV)":<15} {"Half-life"}')
for name, E_val, halflife in nuclei:
    s(f'{name:<10} {E_val:<15} {halflife}')
s('1 MeV difference in E_α → 10²⁰ difference in half-life — Gamow exponential')

sub('Radioactive decay ODE')
t_s = symbols('t', positive=True)
lam = symbols('lambda', positive=True)
N_fn = Function('N')(t_s)
decay_ode = Eq(N_fn.diff(t_s), -lam*N_fn)
s('dN/dt = −λN:', decay_ode)
sol = dsolve(decay_ode)
s('Solution:', sol)
s('Half-life t_{1/2} = ln(2)/λ:', log(2)/lam)
s('Activity A = λN (decays per second)')
```

# §4 — 3D Statistics

```python
sec('3D Statistics')
x, y, z = symbols('x y z', real=True)
mu_x, mu_y, mu_z = symbols('mu_x mu_y mu_z', real=True)
sigma = symbols('sigma', positive=True)

sub('3D Gaussian (isotropic)')
f_3d = (1/(2*pi*sigma**2)**Rational(3,2)) * exp(
    -((x-mu_x)**2 + (y-mu_y)**2 + (z-mu_z)**2)/(2*sigma**2))
s('f(x,y,z) =', f_3d)
s('Normalization: ∫∫∫ f dxdydz = 1')
s('Each dimension independent → joint = product of marginals')

sub('Covariance matrix — general 3D Gaussian')
s('X ~ N(μ, Σ)  where Σ is 3×3 positive definite matrix')
s_xx, s_yy, s_zz = symbols('sigma_xx sigma_yy sigma_zz', positive=True)
s_xy, s_xz, s_yz = symbols('sigma_xy sigma_xz sigma_yz', real=True)
Sigma = Matrix([[s_xx, s_xy, s_xz],
                [s_xy, s_yy, s_yz],
                [s_xz, s_yz, s_zz]])
s('Σ =', Sigma)
mu_vec = Matrix([mu_x, mu_y, mu_z])
r_vec  = Matrix([x, y, z])
diff_vec = r_vec - mu_vec
s('f(r) = (2π)^(-3/2)|Σ|^(-1/2) exp(−½(r−μ)ᵀΣ⁻¹(r−μ))')
s('Exponent:')
exponent = Rational(-1,2) * diff_vec.T * Sigma.inv() * diff_vec
s('−½(r−μ)ᵀΣ⁻¹(r−μ) =', exponent[0,0])

sub('Eigendecomposition of Σ → principal axes')
s('Σ = R D Rᵀ  (R = rotation, D = diagonal variances)')
s('Principal axes = eigenvectors of Σ')
s('In those axes: independent Gaussians along each axis')
s('Ellipsoid of constant probability: (r−μ)ᵀΣ⁻¹(r−μ) = χ²(3, p)')
s('1σ ellipsoid: p = 0.199 (19.9% of probability)')
s('2σ ellipsoid: p = 0.739')
s('3σ ellipsoid: p = 0.971')

sub('Chi-squared distribution in 3D')
r_s = symbols('r', positive=True)
k = 3  # degrees of freedom
s('Radial distance R = |X−μ|/σ follows chi distribution (k=3 dof)')
f_chi = r_s**2 * exp(-r_s**2/2) * sqrt(2/pi)
s('f_R(r) = √(2/π) r² exp(−r²/2)  (Maxwell distribution!)')
s('Mean R = 2√(2/π) ≈ 1.596σ')
s('Mode R = √2 σ')
s('This is the same as Maxwell speed distribution for gas molecules')

sub('3D histogram / kernel density in NumPy')
s('''# 3D KDE estimate
import numpy as np
from scipy.stats import gaussian_kde

# data: shape (3, N)
data = np.random.multivariate_normal([0,0,0],
    [[1,0.5,0],[0.5,1,0.3],[0,0.3,1]], size=10000).T

kde = gaussian_kde(data)

# evaluate on grid
xi = np.linspace(-3,3,30)
grid = np.meshgrid(xi,xi,xi)
pts  = np.vstack([g.ravel() for g in grid])
density = kde(pts).reshape(30,30,30)
''')
```

# §5 — Absolute Value, Noise, Manufacturing Tolerances

```python
sec('Absolute Value in Physics & Manufacturing')
x = symbols('x', real=True)

sub('Absolute value — mathematical properties')
s('|x| = x  if x≥0,  −x  if x<0')
s('|x| = √(x²)', sqrt(x**2))
s('|x| = max(x, −x)')
s('Triangle inequality: |x+y| ≤ |x|+|y|')
s('In complex: |z| = √(z·z*) = √(Re²+Im²)')
s('In physics: |ψ|² = probability density  ← most important use')
s('In signal:  |X(ω)|² = power spectral density')

sub('Manufacturing tolerances — 3σ rule')
s('Part specified as X ± δ')
s('Process has mean μ and std σ')
s('Yield = P(|X−μ| < δ) = erf(δ/σ√2)')
delta, sigma_m = symbols('delta sigma', positive=True)
yield_expr = erf(delta/(sigma_m*sqrt(2)))
s('Yield =', yield_expr)
for n_sig in [1, 2, 3, 6]:
    y = float(erf(n_sig/sqrt(2)).evalf())
    defects_per_million = (1-y)*1e6
    s(f'  {n_sig}σ: yield={y:.4f}  defects={defects_per_million:.1f} ppm')
s('6σ = 3.4 defects per million — semiconductor standard')

sub('Noise in measurements')
s('SNR = signal power / noise power = (A/σ_noise)²')
s('SNR_dB = 10·log₁₀(SNR)')
A, sigma_n = symbols('A sigma_n', positive=True)
SNR = A**2 / sigma_n**2
SNR_dB = 10*log(SNR, 10)
s('SNR =', SNR)
s('SNR_dB =', SNR_dB)
s('Shot noise: σ² = N  (Poisson, counting photons)')
s('Thermal noise: P = k_B T B  (Johnson-Nyquist)')
s('In your GS project: SNR sets minimum |D| needed for convergence')
```
