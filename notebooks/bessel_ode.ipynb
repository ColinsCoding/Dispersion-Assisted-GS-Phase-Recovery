```python
from sympy import *
from IPython.display import display, Markdown
init_printing(use_latex='mathjax')
def s(label, expr=None):
    display(Markdown(f'**{label}**'))
    if expr is not None: display(expr)
def sec(t): display(Markdown(f'---\n# {t}'))
def sub(t): display(Markdown(f'## {t}'))
def row(left_label, left_expr, right_label, right_expr):
    display(Markdown(f'| **{left_label}** | **{right_label}** |'))
    display(Markdown(f'|---|---|'))
    cols = f'| ${latex(left_expr)}$ | ${latex(right_expr)}$ |'
    display(Markdown(cols))
```

# Bessel vs Cartesian — Side by Side
### Every algebra step. Elementary. No gaps.

---
# §1 — The Two ODEs Side by Side

```python
sec('Side by Side: Cartesian vs Bessel')
x, r = symbols('x r')
y = Function('y')
n = symbols('n', nonneg=True)

s('CARTESIAN 2nd order homogeneous:')
cart = Eq(y(x).diff(x,2) + y(x), 0)
s('', cart)
s('BESSEL equation of order n:')
bessel_eq = Eq(r**2*y(r).diff(r,2) + r*y(r).diff(r) + (r**2 - n**2)*y(r), 0)
s('', bessel_eq)

display(Markdown('---'))
display(Markdown('| Cartesian | Bessel |'))
display(Markdown('|---|---|'))
display(Markdown('| y\'\'+ y = 0 | r²y\'\' + ry\' + (r²−n²)y = 0 |'))
display(Markdown('| constant coefficients | variable coefficients |'))
display(Markdown('| try y=eˣ | try y=Σ aₖ rᵏ (power series) |'))
display(Markdown('| solutions: sin(x), cos(x) | solutions: Jₙ(r), Yₙ(r) |'))
display(Markdown('| oscillates on line | oscillates on disk/cylinder |'))
display(Markdown('| boundary: endpoints | boundary: circle radius |'))
```

---
# §2 — Cartesian: Every Algebra Step

```python
sec('Cartesian ODE — y\'\' + y = 0 — Every Step')
x = symbols('x')
y = Function('y')
r_s = symbols('r')

sub('Step 1 — write characteristic equation')
s('Assume y = eʳˣ  →  y\' = reʳˣ  →  y\'\' = r²eʳˣ')
s('Substitute into y\'\' + y = 0:')
s('r²eʳˣ + eʳˣ = 0')
s('eʳˣ(r² + 1) = 0')
s('Since eʳˣ ≠ 0 for any x:')
char_eq = r_s**2 + 1
s('Characteristic equation: r² + 1 = 0', Eq(char_eq, 0))

sub('Step 2 — solve characteristic equation')
s('r² + 1 = 0')
s('r² = −1')
s('r = ±√(−1)')
s('r = ±i  (imaginary roots)')
roots = solve(char_eq, r_s)
s('Roots:', roots)

sub('Step 3 — write general solution')
s('Complex roots r = α ± iβ  →  general solution:')
s('y = eᵅˣ(C₁cos(βx) + C₂sin(βx))')
s('Here α=0, β=1:')
s('y = C₁cos(x) + C₂sin(x)')
sol = dsolve(Eq(y(x).diff(x,2) + y(x), 0))
s('SymPy confirms:', sol)

sub('Step 4 — verify by substitution')
C1, C2 = symbols('C1 C2')
y_sol = C1*cos(x) + C2*sin(x)
ypp   = diff(y_sol, x, 2)
s('y =', y_sol)
s('y\'\'=', ypp)
s('y\'\' + y =', simplify(ypp + y_sol))
s('= 0 ✓')

sub('Step 5 — apply initial conditions')
s('Example: y(0)=1, y\'(0)=0')
eq1 = Eq(y_sol.subs(x,0), 1)
eq2 = Eq(diff(y_sol,x).subs(x,0), 0)
s('y(0)=1:', eq1)
s('y\'(0)=0:', eq2)
s('→ C₁=1, C₂=0  →', solve([eq1,eq2],[C1,C2]))
s('Particular solution: y = cos(x)')
```

# §3 — Bessel: Every Algebra Step

```python
sec('Bessel ODE — Every Step')
r, n_s = symbols('r n', nonneg=True)
y = Function('y')
k, a_k = symbols('k a_k')

sub('Step 1 — where it comes from')
s('Start with Laplace equation ∇²u = 0 in cylindrical coords (r,θ,z):')
s('(1/r)∂/∂r(r ∂u/∂r) + (1/r²)∂²u/∂θ² + ∂²u/∂z² = 0')
s('Separate: u(r,θ,z) = R(r)·Θ(θ)·Z(z)')
s('θ equation → Θ = e^(inθ),  n = integer')
s('z equation → Z = e^(±kz)')
s('r equation becomes (with separation constant k):')
bessel_eq = Eq(r**2*y(r).diff(r,2) + r*y(r).diff(r) + (r**2 - n_s**2)*y(r), 0)
s('', bessel_eq)
s('This is Bessel\'s equation of order n')

sub('Step 2 — why we can\'t use eʳ trick')
s('Coefficient of y\'\'is r², NOT a constant')
s('eʳ approach → eʳ(r² + r + r² − n²) = 0 → not solvable simply')
s('Must use POWER SERIES (Frobenius method)')

sub('Step 3 — Frobenius method: assume y = Σ aₖ r^(k+s)')
s('y = Σₖ₌₀^∞ aₖ r^(k+s)')
s('y\' = Σₖ (k+s) aₖ r^(k+s-1)')
s('y\'\' = Σₖ (k+s)(k+s-1) aₖ r^(k+s-2)')
s('Substitute into Bessel equation:')
s('Σₖ (k+s)(k+s-1)aₖ r^(k+s) + Σₖ (k+s)aₖ r^(k+s) + Σₖ aₖ r^(k+s+2) − n²Σₖ aₖ r^(k+s) = 0')

sub('Step 4 — indicial equation (k=0 term)')
s('Collect r^s terms (k=0):')
s('[s(s-1) + s − n²] a₀ = 0')
s('s² − n²  = 0')
s('s = ±n     ← indicial equation')
s('Take s = +n (gives Jₙ, the well-behaved solution)')

sub('Step 5 — recurrence relation')
s('Collecting r^(k+n) terms:')
s('[(k+n)² − n²] aₖ + aₖ₋₂ = 0')
s('k(k+2n) aₖ = −aₖ₋₂')
s('aₖ = −aₖ₋₂ / [k(k+2n)]')
s('Odd coefficients: a₁=0 → a₃=a₅=...=0')
s('Even coefficients: k=2m:')
s('a₂ₘ = (−1)ᵐ a₀ / [2²ᵐ m! (n+m)!] · n!')

sub('Step 6 — Bessel function Jₙ(r)')
s('Choose a₀ = 1/(2ⁿ n!):')
s('Jₙ(r) = Σₘ₌₀^∞ (−1)ᵐ / [m!(m+n)!] · (r/2)^(2m+n)')
m_s = symbols('m', nonneg=True, integer=True)
J_series_term = ((-1)**m_s / (factorial(m_s)*factorial(m_s+n_s))) * (r/2)**(2*m_s+n_s)
s('Term m:', J_series_term)

sub('Step 7 — SymPy Bessel functions')
from sympy import besselj, bessely, besseli, besselk
for n_val in [0, 1, 2]:
    J = besselj(n_val, r)
    Y = bessely(n_val, r)
    s(f'J_{n_val}(r) =', J)
    s(f'Y_{n_val}(r) =', Y)
    ser = series(besselj(n_val, r), r, 0, 8)
    s(f'J_{n_val}(r) series =', ser)
```

# §4 — Side-by-Side Comparison Table

```python
sec('Complete Side-by-Side')
x, r = symbols('x r', positive=True)

display(Markdown('''
| Property | Cartesian y\'\'+y=0 | Bessel r²y\'\'+ry\'+(r²-n²)y=0 |
|---|---|---|
| Geometry | Line (1D) | Disk / cylinder (2D/3D) |
| Coefficients | Constant | Variable (r², r) |
| Solution method | Characteristic eq | Frobenius power series |
| Solution 1 | cos(x) | Jₙ(r) (finite at r=0) |
| Solution 2 | sin(x) | Yₙ(r) (blows up at r=0) |
| Behavior | Uniform oscillation | Decaying oscillation |
| Zeros | x = nπ or (n+½)π | Not equally spaced |
| Orthogonality | ∫cos(mx)cos(nx)dx=δ | ∫Jₙ(αr)Jₙ(βr)r dr=δ |
| Completeness | Fourier series | Fourier-Bessel series |
| Physical use | Strings, boxes | Drums, waveguides, atoms |
| Boundary cond | y(0)=y(L)=0 | finite at r=0, Jₙ(αR)=0 |
''')
)

sub('First zeros of J₀, J₁, J₂ (needed for boundary conditions)')
from sympy import nsolve
for n_val in [0, 1, 2]:
    s(f'Zeros of J_{n_val}(r):')
    zeros = []
    guesses = [2.4, 5.5, 8.6, 11.8] if n_val==0 else \
              [3.8, 7.0, 10.2, 13.3] if n_val==1 else \
              [5.1, 8.4, 11.6, 14.8]
    for g in guesses:
        try:
            z = float(nsolve(besselj(n_val, r), r, g))
            zeros.append(f'{z:.4f}')
        except:
            pass
    s(f'  first 4 zeros: {zeros}')
```

# §5 — Physical Example: Drum Head

```python
sec('Physical Example: Vibrating Drum Head')
r, theta, t_s = symbols('r theta t', positive=True)
c, R_drum = symbols('c R', positive=True)

sub('PDE: wave equation in 2D polar')
s('∂²u/∂t² = c²∇²u = c²[(1/r)∂/∂r(r∂u/∂r) + (1/r²)∂²u/∂θ²]')
s('Boundary: u(R,θ,t) = 0  (edge clamped)')
s('Separate: u(r,θ,t) = R(r)·Θ(θ)·T(t)')

sub('Each separated equation')
s('Time:    T\'\'= −ω²T       → T = cos(ωt) or sin(ωt)')
s('Angle:   Θ\'\' = −m²Θ      → Θ = cos(mθ) or sin(mθ),  m=0,1,2,...')
s('Radial:  r²R\'\'+ rR\' + (k²r²−m²)R = 0  ← Bessel equation!')
s('Solution: R(r) = Jₘ(kr)')
s('Boundary R(R_drum)=0: Jₘ(kR) = 0  → k = αₘₙ/R')
s('where αₘₙ = nth zero of Jₘ')

sub('Normal mode frequencies')
s('ω_{mn} = c·αₘₙ/R')
s('u_{mn}(r,θ,t) = Jₘ(αₘₙ r/R)·cos(mθ)·cos(ω_{mn}t)')
s('Mode (m=0,n=1): symmetric, one nodal circle')
s('Mode (m=1,n=1): one nodal diameter')
s('Mode (m=0,n=2): two nodal circles')

sub('Compare: 1D string vs 2D drum')
display(Markdown('''
| | 1D String | 2D Drum |
|---|---|---|
| ODE | y\'\' + k²y = 0 | Bessel equation |
| Solutions | sin(nπx/L) | Jₘ(αₘₙ r/R) |
| Frequencies | ω_n = nπc/L | ω_mn = c·αₘₙ/R |
| Harmonic? | Yes: ω_n = n·ω₁ | No: not integer multiples |
| Sound | Musical (harmonics) | Non-musical (drum!) |
'''))
```

# §6 — General 2nd Order Homogeneous: All Cases

```python
sec('2nd Order Homogeneous ODE — All Cases')
x = symbols('x')
y = Function('y')
a, b, c_s = symbols('a b c', real=True)
r_s = symbols('r')

sub('ay\'\'+ by\' + cy = 0  →  ar² + br + c = 0')
char = a*r_s**2 + b*r_s + c_s
disc = b**2 - 4*a*c_s
s('Discriminant Δ = b²−4ac:', disc)

cases = [
    ('Δ > 0: two distinct real roots r₁, r₂',
     'y = C₁e^(r₁x) + C₂e^(r₂x)',
     Eq(y(x).diff(x,2) - 5*y(x).diff(x) + 6*y(x), 0)),
    ('Δ = 0: repeated root r = −b/2a',
     'y = (C₁ + C₂x)e^(rx)',
     Eq(y(x).diff(x,2) - 4*y(x).diff(x) + 4*y(x), 0)),
    ('Δ < 0: complex roots α ± iβ',
     'y = e^(αx)(C₁cos(βx) + C₂sin(βx))',
     Eq(y(x).diff(x,2) - 2*y(x).diff(x) + 5*y(x), 0)),
]

for desc, form, ode in cases:
    s(f'── {desc} ──')
    s('Form:', Markdown(f'`{form}`'))
    s('Example ODE:', ode)
    char_ex = ode.lhs.as_poly(y(x))
    sol = dsolve(ode)
    s('Solution:', sol)
    s('Verify (substitute back):', simplify(ode.subs(y(x), sol.rhs).doit()))

sub('Euler-Cauchy: ax²y\'\' + bxy\' + cy = 0  →  try y = rˣ')
s('Substitute y = xⁿ:')
s('an(n-1)xⁿ + bnxⁿ + cxⁿ = 0')
s('an(n-1) + bn + c = 0  ← indicial equation')
n_s = symbols('n')
a_e, b_e, c_e = 1, -3, 4
indicial = a_e*n_s*(n_s-1) + b_e*n_s + c_e
s(f'Example x²y\'\' − 3xy\' + 4y = 0  →  n(n-1) − 3n + 4 = 0')
s('Indicial:', Eq(indicial, 0))
roots_e = solve(indicial, n_s)
s('Roots:', roots_e)
ode_ec = Eq(x**2*y(x).diff(x,2) - 3*x*y(x).diff(x) + 4*y(x), 0)
s('Solution:', dsolve(ode_ec))
```
