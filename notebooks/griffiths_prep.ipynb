# §1 — Logarithmic Differentiation

```python
sec('Logarithmic Differentiation')
x = symbols('x', positive=True)

sub('When to use: products, quotients, variable exponents')
s('Rule: if y=f(x), take ln both sides → ln y = ln f(x)')
s('Differentiate → (1/y)·dy/dx = d/dx[ln f]')
s('Multiply both sides by y → dy/dx = y · d/dx[ln f]')

sub('Case 1 — variable exponent: y = xˣ')
f1 = x**x
s('y = xˣ')
s('ln y = x ln x')
s('(1/y)y\' = ln x + 1')
s('y\' = xˣ(ln x + 1):', diff(f1, x))

sub('Case 2 — messy product: y = x²sin(x)eˣ')
f2 = x**2 * sin(x) * exp(x)
s('y =', f2)
s('ln y = 2 ln x + ln sin x + x')
s('(1/y)y\' = 2/x + cos x/sin x + 1')
s('y\' =', diff(f2, x))
s('Factored:', factor(diff(f2, x)))

sub('Case 3 — variable base AND exponent: y = (sin x)^(ln x)')
f3 = sin(x)**log(x)
s('y =', f3)
s('ln y = ln(x)·ln(sin x)')
s('y\' =', simplify(diff(f3, x)))

sub('Case 4 — rational with square roots')
f4 = sqrt(x*(x+1)) / (x+2)**3
s('y =', f4)
s('ln y = ½ln x + ½ln(x+1) − 3ln(x+2)')
s('y\' =', simplify(diff(f4, x)))
```

# §2 — Related Rates: Full Method

```python
sec('Related Rates — Full Method')
t = symbols('t')

sub('Recipe — works every time')
s('1. Identify all quantities changing with time')
s('2. Write ONE equation relating them (geometry or physics)')
s('3. Differentiate both sides with respect to t')
s('4. Plug in known values, solve for unknown rate')
s('NEVER plug in values before differentiating')

sub('Example 1 — two ships diverging')
x_fn = Function('x')(t)
y_fn = Function('y')(t)
D    = Function('D')(t)
s('Ship A moves east at 20 km/h, Ship B moves north at 15 km/h')
s('How fast is distance between them increasing after 2 hours?')
s('Equation: D² = x² + y²')
rel = Eq(D**2, x_fn**2 + y_fn**2)
s('Differentiate:', Eq(2*D*Derivative(D,t), 2*x_fn*Derivative(x_fn,t) + 2*y_fn*Derivative(y_fn,t)))
s('At t=2: x=40, y=30, D=50, dx/dt=20, dy/dt=15')
dDdt = (40*20 + 30*15) / 50
s('dD/dt =', dDdt, )

sub('Example 2 — light on wall (angle rate)')
s('Lighthouse rotates at 2 rev/min. Wall 1km away.')
s('How fast does light spot move when beam at 45°?')
theta_fn = Function('theta')(t)
x_fn2 = Function('x')(t)
s('x = tan θ  (normalized, wall at distance 1)')
s('dx/dt = sec²θ · dθ/dt')
omega_val = 2*pi*2  # 2 rev/min in rad/min
theta_val = pi/4
dxdt = (1/cos(theta_val)**2) * omega_val
s(f'At θ=45°: dx/dt = sec²(π/4)·4π = {float(dxdt):.2f} km/min')
```

# §3 — Vector Calculus for Griffiths

```python
sec('Vector Calculus — Everything Griffiths Uses')
C = CoordSys3D('C')
x, y, z = C.x, C.y, C.z
i, j, k = C.i, C.j, C.k

sub('The operators — one line each')
s('∇f   = gradient    → vector pointing uphill')
s('∇·F  = divergence  → source/sink density')
s('∇×F  = curl        → rotation density')
s('∇²f  = Laplacian   → ∇·(∇f), average deviation from neighbors')

sub('Key identities Griffiths uses constantly')
r_vec = x*i + y*j + z*k
r_mag = sqrt(x**2 + y**2 + z**2)
phi   = x**2 + y**2 + z**2  # r²

identities = [
    ('∇(r²)',        gradient(phi, C)),
    ('∇·r',          divergence(r_vec, C)),
    ('∇×r',          curl(r_vec, C)),
    ('∇²(r²)',       laplacian(phi, C)),
]
for label, val in identities:
    s(f'{label} =', val)

sub('∇(1/r) — appears in every Coulomb problem')
inv_r = 1/r_mag
grad_inv_r = gradient(inv_r, C)
s('∇(1/r) =', grad_inv_r)
s('Simplified: −r̂/r²  ← Coulomb field direction')

sub('Divergence theorem: ∯F·da = ∭(∇·F)dV')
s('Used in Griffiths to derive Gauss law in differential form')
s('∯E·da = Q_enc/ε₀  →  ∭(∇·E)dV = ∭(ρ/ε₀)dV  →  ∇·E = ρ/ε₀')

sub('Stokes theorem: ∮F·dl = ∬(∇×F)·da')
s('Used to derive Faraday and Ampere in differential form')
s('∮E·dl = −dΦ_B/dt  →  ∇×E = −∂B/∂t')

sub('Dirac delta — Griffiths section 1.5')
s('∇·(r̂/r²) = 4π δ³(r)  ← most important identity in Griffiths')
s('Without δ: ∇·(r̂/r²) = 0 for r≠0  BUT integral over any sphere = 4π')
s('δ³(r) captures the singularity at origin')
s('∭ δ³(r−r\') f(r) dV = f(r\')  ← sifting property')
s('Coulomb: ∇²(1/r) = −4π δ³(r)')
s('This is WHY Gauss law works for point charges')
```

# §4 — Linear Algebra for Quantum Mechanics

```python
sec('Linear Algebra → Quantum Mechanics')

sub('The translation table')
table = [
    ('Vector v',           'State |ψ⟩'),
    ('Basis eᵢ',           'Eigenstates |n⟩'),
    ('Component vᵢ=eᵢ·v', 'Amplitude cₙ=⟨n|ψ⟩'),
    ('Dot product',        'Inner product ⟨φ|ψ⟩=∫φ*ψ dx'),
    ('Matrix A',           'Operator Â'),
    ('Eigenvalue Av=λv',   'Âψ=aψ  (measurement outcome = a)'),
    ('Hermitian Aᵀ=A',     'Observable Â†=Â'),
    ('Unitary UᵀU=I',      'Time evolution Û†Û=I'),
    ('Trace Σᵢ Aᵢᵢ',       'Expectation ⟨A⟩=⟨ψ|Â|ψ⟩'),
]
for linalg, qm in table:
    s(f'{linalg:<30} ↔  {qm}')

sub('Hermitian operators — why observables must be Hermitian')
s('Eigenvalues of Hermitian operator are REAL')
s('Proof: Â|n⟩=aₙ|n⟩  →  ⟨n|Â|n⟩=aₙ')
s('Also ⟨n|Â|n⟩=⟨Â†n|n⟩=⟨Ân|n⟩=aₙ*')
s('So aₙ=aₙ*  →  aₙ is real ✓')
s('Eigenvectors of Hermitian op with different eigenvalues are orthogonal')
A = Matrix([[2, 1+I], [1-I, 3]])
s('Example Hermitian matrix:', A)
s('A† =', A.H)
s('A†=A:', A.H == A)
eigs = A.eigenvects()
for val, mult, vecs in eigs:
    s(f'λ={val}:', vecs[0].T)

sub('Commutator [A,B] = AB−BA')
X_op, P_op = symbols('X P')
s('[X̂, P̂] = iħ  ← canonical commutation relation')
s('Non-zero commutator → operators cannot be simultaneously measured')
s('Uncertainty principle: σ_A σ_B ≥ |⟨[A,B]⟩|/2')
s('For X,P: σ_x σ_p ≥ ħ/2')
s('')
hbar = symbols('hbar', positive=True)
s('[X̂², P̂] = X̂[X̂,P̂] + [X̂,P̂]X̂ = 2iħX̂')
s('[X̂, P̂²] = 2iħP̂  (by same method)')
s('[X̂ⁿ, P̂] = iħnX̂^(n-1)')
s('[f(X̂), P̂] = iħ df/dX̂')
```

# §5 — Calculus Tools Griffiths Assumes

```python
sec('Calculus Griffiths Assumes You Know')
x, k, a, n_s = symbols('x k a n', positive=True)

sub('Gaussian integrals — used in every QM calculation')
gaussian_ints = [
    ('∫₋∞^∞ e^(−ax²) dx',       integrate(exp(-a*x**2), (x,-oo,oo))),
    ('∫₋∞^∞ x² e^(−ax²) dx',    integrate(x**2*exp(-a*x**2), (x,-oo,oo))),
    ('∫₋∞^∞ x⁴ e^(−ax²) dx',    integrate(x**4*exp(-a*x**2), (x,-oo,oo))),
    ('∫₀^∞ e^(−ax²) dx',         integrate(exp(-a*x**2), (x,0,oo))),
    ('∫₀^∞ x e^(−ax²) dx',       integrate(x*exp(-a*x**2), (x,0,oo))),
    ('∫₀^∞ x² e^(−ax) dx',       integrate(x**2*exp(-a*x), (x,0,oo))),
]
for label, val in gaussian_ints:
    s(f'{label} =', val)

sub('Integration by parts — IBP for QM matrix elements')
s('∫ψ* d²φ/dx² dx = [ψ* dφ/dx]₋∞^∞ − ∫(dψ*/dx)(dφ/dx) dx')
s('Boundary terms vanish for normalizable wavefunctions')
s('→ ∫ψ* Hφ dx = ∫(Hψ)* φ dx  (H is Hermitian)')

sub('Fourier transforms — momentum space')
p, hbar_s = symbols('p hbar', positive=True)
s('φ(p) = (1/√2πħ) ∫ ψ(x) e^(−ipx/ħ) dx')
s('ψ(x) = (1/√2πħ) ∫ φ(p) e^(+ipx/ħ) dp')
s('|φ(p)|² dp = probability of finding momentum in [p, p+dp]')
s('Parseval: ∫|ψ|²dx = ∫|φ|²dp  (normalization preserved)')

sub('Spherical harmonics Yₗᵐ — Griffiths chapter 4')
theta, phi_s = symbols('theta phi', real=True)
s('Solutions to angular part of Schrödinger in spherical coords')
s('∇²ψ = (1/r²)∂/∂r(r²∂ψ/∂r) + (1/r²sinθ)∂/∂θ(sinθ∂ψ/∂θ) + (1/r²sin²θ)∂²ψ/∂φ²')
s('Separation: ψ(r,θ,φ) = R(r)·Y(θ,φ)')
s('')
s('First few Yₗᵐ(θ,φ):')
from sympy import Ynm
for l_val in range(3):
    for m_val in range(-l_val, l_val+1):
        Y = Ynm(l_val, m_val, theta, phi_s)
        s(f'Y_{l_val}^{m_val} =', Y.expand(func=True))

sub('Legendre polynomials Pₗ(cos θ) — l=0,1,2,3')
x_leg = symbols('x')
for l_val in range(4):
    P = legendre(l_val, x_leg)
    s(f'P_{l_val}(x) =', P)
```

# §6 — Statistics for Quantum Mechanics

```python
sec('Statistics — Expectation Values and Uncertainty')
x, sigma, mu = symbols('x sigma mu', real=True)

sub('Expectation value')
s('Discrete: ⟨j⟩ = Σⱼ j·P(j)')
s('Continuous: ⟨x⟩ = ∫ x·|ψ(x)|² dx')
s('Operator: ⟨Q⟩ = ∫ ψ* Q̂ ψ dx')

sub('Standard deviation and variance')
s('σ² = ⟨x²⟩ − ⟨x⟩²  ← always use this form')
s('σ = √(⟨x²⟩ − ⟨x⟩²)')
s('Prove: σ² = ⟨(x−⟨x⟩)²⟩ = ⟨x²⟩ − 2⟨x⟩⟨x⟩ + ⟨x⟩² = ⟨x²⟩ − ⟨x⟩²')

sub('Worked: particle in box ground state')
x_s, L = symbols('x L', positive=True)
psi = sqrt(2/L)*sin(pi*x_s/L)
s('ψ₁(x) = √(2/L) sin(πx/L)')
x_exp   = integrate(x_s * psi**2, (x_s, 0, L))
x2_exp  = integrate(x_s**2 * psi**2, (x_s, 0, L))
sigma_x = sqrt(simplify(x2_exp - x_exp**2))
s('⟨x⟩ =', simplify(x_exp))
s('⟨x²⟩ =', simplify(x2_exp))
s('σ_x =', simplify(sigma_x))

hbar = symbols('hbar', positive=True)
m, L2 = symbols('m L', positive=True)
p_exp  = 0  # by symmetry (odd integrand)
p2_exp = (pi*hbar/L2)**2
sigma_p = sqrt(p2_exp)
s('⟨p⟩ = 0  (odd integrand on symmetric domain)')
s('⟨p²⟩ = (πħ/L)²:', p2_exp)
s('σ_p = πħ/L')
product = simplify(sigma_x.subs(L, L2) * sigma_p)
s('σ_x · σ_p =', product)
s('ħ/2 =', hbar/2)
s('σ_x σ_p > ħ/2 ✓  uncertainty principle satisfied')
```

# §7 — Rational Algebra Moves in Calculus

```python
sec('Rational Algebra — Moves That Show Up in Calculus')
x, a, b, n_s = symbols('x a b n')

sub('Completing the square — used in Gaussian integrals and QM')
for expr in [x**2 + 6*x + 5, 2*x**2 - 4*x + 7, x**2 + b*x]:
    completed = complete_the_square(expr, x) if False else None
    s(f'{expr} =', sqf(expr))
    s('  complete the square:', simplify(expr - (x + b/2)**2 + b**2/4) if b in expr.free_symbols else
      simplify(expr))

sub('Partial fractions — every Laplace/integral problem')
fracs = [
    1/((x+1)*(x+2)),
    (x+3)/((x-1)*(x**2+1)),
    x**2/((x+1)*(x-1)**2),
]
for f in fracs:
    pf = apart(f, x)
    s(f'{f} =', pf)

sub('Rationalization — used in limits and trig substitution')
cases = [
    ('1/(√x+1) · (√x−1)/(√x−1)', 1/(sqrt(x)+1)),
    ('(√(x+h)−√x)/h → 0/0 form',  (sqrt(x + symbols('h'))-sqrt(x))/symbols('h')),
]
for label, expr in cases:
    s(label)
    s('  rationalized:', radsimp(expr))

sub('Common factoring patterns in physics')
patterns = [
    ('difference of squares', x**2 - a**2,       factor(x**2 - a**2)),
    ('sum of cubes',          x**3 + a**3,        factor(x**3 + a**3)),
    ('difference of cubes',   x**3 - a**3,        factor(x**3 - a**3)),
    ('perfect square',        x**2 + 2*a*x + a**2, factor(x**2 + 2*a*x + a**2)),
    ('factor out r²',         a**2*x**2 + a**2*b**2, factor(a**2*x**2 + a**2*b**2)),
]
for name, expr, factored in patterns:
    s(f'{name}: {expr} =', factored)
```
